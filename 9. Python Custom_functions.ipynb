* **def**: This keyword indicates the start of a function definition. It stands for "define"



* **Function_name**: This is the name of the function you are defining.

* **Parameters**: These are the input values that the function will take. They are optional, and you can have as many or as few as you need.

* **Parameter Hints** : Pressing "Shift+Tab" while calling a function will display the parameter hints that were initially defined.

* **Default Parameter Value** : If a value is not provided when calling a function, this value will be used instead.

* **Docstring** : A docstring is a string literal that appears as the first statement in a module, function, class, or method definition. Its purpose is to provide documentation for the code and to describe what the function does, what arguments it takes, and what it returns.

* **Function code** : This is the code that gets executed when the function is called. It can be any valid Python code.

* **return**: This keyword indicates the value that the function will return. This is also optional, and you can have multiple return statements in a function.

```python
# Program to check if a number is prime or not

num = 29

# To take input from the user
#num = int(input("Enter a number: "))

# define a flag variable
flag = False

if num == 0 or num == 1:
    print(num, "is not a prime number")
elif num > 1:
    # check for factors
    for i in range(2, num):
        if (num % i) == 0:
            # if factor is found, set flag to True
            flag = True
            # break out of loop
            break

    # check if flag is True
    if flag:
        print(num, "is not a prime number")
    else:
        print(num, "is a prime number")
```

```python
def check_prime():
  num = int(input("Enter a number : "))
  if num <= 1:
    print(num, "is NOT a prime number.")
    return
  for i in range(2, num):
    if num % i == 0:
      print(num, "is Not a prime number.")
      return
  print(num, "is a prime number.")
  return
```

```python
check_prime()
```

# `*args` and `**kwargs`

*   `*args` is used to pass a non-keyworded, variable-length argument list to a function.
*   `**kwargs` is used to pass a keyworded, variable-length argument list to a function.

```python
help(my_sum)
```

```python
def my_sum(x,y=10):
  """Calculates the sum of a variable number of arguments."""
  total = x+y
  return total
```

```python
def my_sum(*args):
  """Calculates the sum of a variable number of arguments."""
  total = 0
  for x in args:
    total += x
  return total

print(my_sum(1, 2, 3, 4,100,200))
print(my_sum(10, 20, 30, 40, 50))
```

```python
def my_function(**kwargs):
  """Prints key-value pairs from keyword arguments."""
  for key, value in kwargs.items():
    print(f"{key}: {value}")

my_function(a=1, b=2, c=3, h=5)
my_function(name="Alice", city="London",pincode="123455",mobile=12345678)
```

```python
# Get the year as input from the user and convert it to an integer
year = int(input())

# Get the month as input from the user and convert it to an integer
month = int(input())

# Get the day as input from the user and convert it to an integer
day = int(input())

# Check if the year is a leap year
# A leap year is divisible by 400, or divisible by 4 but not 100
if (year % 400 == 0):
  leap_year = True  # If divisible by 400, it's a leap year
elif (year % 100 == 0):
  leap_year = False  # If divisible by 100 but not 400, not a leap year
elif (year % 4 == 0):
  leap_year = True  # If divisible by 4 but not 100, it's a leap year
else:
  leap_year = False  # Otherwise, it's not a leap year

# Determine the number of days in the given month
if month in (1, 3, 5, 7, 8, 10, 12):
  month_length = 31  # Months with 31 days
elif month == 2:
  if leap_year:
    month_length = 29  # February in a leap year has 29 days
  else:
    month_length = 28  # February in a common year has 28 days
else:
  month_length = 30  # Other months have 30 days

# Check if it's possible to increment the day
if day < month_length:
  day += 1  # If the current day is less than the max days in the month, increment the day
else:
  # If it's the last day of the month
  day = 1  # Reset the day to 1 (start of the next month)
  if month == 12:
    # If it's December, move to the next year
    month = 1  # Reset month to January
    year += 1  # Increment the year by 1
  else:
    month += 1  # Otherwise, just move to the next month

# Print the next date in the format [yyyy-mm-dd]
print ("The next date is [yyyy-mm-dd] %d-%d-%d." % (year, month, day))
```
