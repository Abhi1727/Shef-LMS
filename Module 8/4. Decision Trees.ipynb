## **What is a Decision Tree?**

<b>Decision tree </b> is a type of <b>supervised learning algorithm</b> that is mostly used in classification problems. It works for both categorical and continuous input and output variables.

### **How to split?**

The decision of making strategic splits heavily affects a tree’s accuracy. The decision criteria is different for classification and regression trees.

Decision trees use multiple algorithms to decide to split a node in two or more sub-nodes. The creation of sub-nodes increases the homogeneity of resultant sub-nodes. In other words, we can say that purity of the node increases with respect to the target variable. Decision tree splits the nodes on all available variables and then selects the split which results in most homogeneous sub-nodes.

<b>Methods to determine best split:</b>

1. Information Gain

2. Gini

3. Chi Square

4. Reduction in Variance



#### 1. Information Gain

Look at the image below and think which node can be described easily. I am sure, your answer is C because it requires less information as all values are similar. On the other hand, B requires more information to describe it and A requires the maximum information. In other words, we can say that C is a Pure node, B is less Impure and A is more impure.

Screenshot 2020-12-16 at 4.07.53 PM.png

Now, we can build a conclusion that less impure node requires less information to describe it. And, more impure node requires more information. Information theory tries to measure and define this degree of disorganization in a system known as *Entropy*. If the sample is completely homogeneous, then the entropy is zero and if the sample is an equally divided (50% – 50%), it has entropy of one.

Entropy can be calculated using formula:-

Here p and q is probability of success and failure respectively in that node. Entropy is also used with categorical target variable. It chooses the split which has lowest entropy compared to parent node and other splits. The lesser the entropy, the better it is.

The decrease is entropy after split is called **Information Gain**.
Steps to calculate information split for a split:

1. Calculate entropy of parent node
2. Calculate entropy of each individual node of split and calculate weighted average of all sub-nodes available in split.
3. Calculate the difference in entropy before and after split.

<b>Example:</b> – Referring to example used above, where we want to segregate the students based on target variable ( playing cricket or not ). In the snapshot below, we split the population using two input variables Gender and Class. Now, I want to identify which split is producing more homogeneous sub-nodes using Information Gain.

gini_index.png

Entropy for parent node = - (15/30) log(15/30) – (15/30) log(15/30) = 1.

Here 1 shows that it is a impure node.

**Split on Gender:**

Entropy for Female node = -(2/10) log(2/10) – (8/10) log(8/10) = 0.72 and for male node,  -(13/20) log(13/20) – (7/20) log(7/20) = 0.93


Entropy for split Gender = Weighted entropy of sub-nodes = (10/30)*0.72 + (20/30)*0.93 = 0.86

Information Gain for split Gender = Entropy before split - Entropy after split = 1 - 0.86 = 0.14

**Split on Class:**

Entropy for Class IX node, -(6/14) log(6/14) – (8/14) log(8/14) = 0.99 and for Class X node,  -(9/16) log(9/16) – (7/16) log(7/16) = 0.99.


Entropy for split Class =  (14/30)*0.99 + (16/30)*0.99 = 0.99

Information Gain for split Class = Entropy before split - Entropy after split = 1 - 0.99 = 0.01

**Decision:**

Above, we can see that *Information Gain* for split on *Gender* is the highest among all, so the tree will split on *Gender*.

#### 2. Gini

Gini  says, if we select two items from a population at random then, if the population is pure, they must be of same class and probability for this is 1.

1. It works with categorical target variable “Success” or “Failure”.
2. It performs only binary splits.
3. Higher the value of Gini higher the homogeneity.
4. CART (Classification and Regression Tree) uses Gini method to create binary splits.
    
Steps to calculate Gini for a split

1. Calculate Gini for sub-nodes, using formula sum of square of probability for success and failure <b>(p^2+q^2).</b>
2. Calculate Gini for split using weighted Gini score of each node of that split

Let’s use Gini method to identify best split for student example.

Split on Gender:

<I>Calculate, Gini for sub-node Female</I> = (0.2)*(0.2)+(0.8)*(0.8)=0.68

<I>Gini for sub-node Male </I>= (0.65)*(0.65)+(0.35)*(0.35)=0.55

<I>Calculate weighted Gini for Split Gender </I>= (10/30)*0.68+(20/30)*0.55 = 0.59

Similar for Split on Class:

<I>Gini for sub-node Class IX</I> = (0.43)*(0.43)+(0.57)*(0.57)=0.51

<I>Gini for sub-node Class X </I>= (0.56)*(0.56)+(0.44)*(0.44)=0.51

<I>Calculate weighted Gini for Split Class </I>= (14/30)*0.51+(16/30)*0.51 = 0.51


Above, you can see that Gini score for split on *Gender* is higher than split on *Class*, hence, the node split will take place on *Gender*.

You might often come across the term ‘Gini Impurity’ which is determined by subtracting the gini value from 1. So mathematically we can say,

Gini Impurity  = 1 - Gini

### **Implementing a Simple Decision Tree:**

A Decision Tree is a type of machine learning algorithm that is used for both classification and regression tasks. It's like a flowchart that makes decisions based on the features of the data. Each node in the tree represents a test on an attribute (e.g., whether a fruit is red), each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes).

Let's consider a simple example using the Decision Tree Classifier in Python. We'll use a small dataset of fruits characterized by their color and size to determine whether they are apples or bananas.

### **Best Practices:**

When working with decision trees in machine learning, adhering to best practices is crucial for achieving optimal performance and accurate results. Here are some best practices to consider:

1. **Data Preprocessing**:
   - **Handle Missing Values**: Ensure that missing values in your dataset are appropriately handled, either by imputation or removal.
   - **Feature Engineering**: Engage in thoughtful feature selection and engineering to improve model performance.

2. **Avoid Overfitting**:
   - **Pruning**: Use pruning techniques to trim the tree and remove sections that may lead to overfitting.
   - **Limit Tree Depth**: Set a maximum depth for the tree to prevent it from becoming overly complex.

3. **Use Ensemble Methods**:
   - Consider using ensemble methods like Random Forests or Gradient Boosting, which combine multiple decision trees to improve performance and robustness.

4. **Splitting Criteria**:
   - **Choose Appropriate Criteria**: Select an appropriate splitting criterion (like Gini impurity or Information Gain) based on your specific problem and data characteristics.

5. **Handling Imbalanced Data**:
   - If dealing with imbalanced datasets, use techniques like SMOTE (Synthetic Minority Over-sampling Technique) or adjust class weights to ensure the model doesn't become biased towards the majority class.

6. **Cross-Validation**:
   - Employ cross-validation techniques to assess the performance of your decision tree and ensure its ability to generalize to unseen data.

7. **Interpretability**:
   - Leverage the interpretability of decision trees. They offer a clear visualization of the decision-making process, which can be crucial in certain applications.

8. **Parameter Tuning**:
   - Experiment with different parameters (like minimum samples split, minimum samples leaf) to find the optimal configuration for your model.

9. **Feature Scaling**:
   - While decision trees are generally not sensitive to the scale of features, ensuring consistent scales can be beneficial, especially when comparing feature importance.

10. **Validation Metrics**:
    - Choose appropriate validation metrics (like accuracy, precision, recall, F1-score) based on the nature of your problem and the balance of classes in your dataset.

By following these best practices, you can maximize the effectiveness of decision trees in your machine learning projects, ensuring they are well-tuned, robust, and suitable for your specific data and problem.

### **Shortcomings:**

Despite their versatility and popularity in machine learning, decision trees come with certain shortcomings that are important to consider:

1. **Overfitting**: Decision trees are prone to overfitting, especially if they are allowed to grow without constraints. An overfitted tree may perform well on training data but poorly on unseen data.

2. **Sensitivity to Data Variations**: Small changes in the data can lead to significantly different tree structures. This high variance makes decision trees less reliable for predictive modeling compared to other algorithms.

3. **Handling of Continuous Variables**: Decision trees can struggle with continuous variables, often leading to complex trees that overfit. They typically perform better with categorical variables.

4. **Handling of Imbalanced Data**: Decision trees can become biased towards the majority class in imbalanced datasets, leading to poor performance in predicting the minority class.

5. **Complexity with Large Datasets**: Building decision trees on large datasets can become computationally expensive and time-consuming. The complexity increases with the number of features and data points.

6. **Linear Relationships**: Decision trees are not the best choice for capturing linear relationships among variables. Linear models or other algorithms might be more suitable in such scenarios.

7. **Limited Interpretability with Large Trees**: While small decision trees are easy to interpret, large trees with numerous splits can become complex and difficult to interpret.

8. **Predictive Accuracy**: In isolation, decision trees often do not have the same level of predictive accuracy as other more advanced algorithms, especially in complex tasks.

9. **No Support for Online Learning**: Decision trees do not support online learning, meaning they cannot be incrementally updated and must be rebuilt from scratch when new data comes in.

10. **Handling of Missing Values**: Decision trees can be sensitive to missing data, and how missing values are handled can significantly impact the performance of the tree.

Understanding these limitations is crucial when choosing decision trees for a machine learning problem. Often, ensemble methods like Random Forests or Gradient Boosting, which combine multiple decision trees, are used to overcome some of these shortcomings.

### **Important Terminology**

Let’s look at the basic terminology used to study decision trees:

terminology.png



<b>Root Node:</b> It represents entire population or sample and this further gets divided into two sets.

<b>Splitting:</b> It is a process of dividing a node into two sub-nodes.

<b>Decision Node:</b> When a sub-node splits into further sub-nodes, then it is called decision node.

<b> Leaf/ Terminal Node:</b> Nodes do not split is called leaf or terminal node.

<b> Pruning:</b> When we remove sub-nodes of a decision node, this process is called pruning. You can say opposite process of splitting.

<b>Branch / Sub-Tree:</b> A sub section of entire tree is called branch or sub-tree.

<b>Parent and Child Node:</b> A node, which is divided into sub-nodes is called parent node of sub-nodes where as sub-nodes are the child of parent node.


### **Introduction to Ensemble Learning:**

Ensemble learning is a powerful concept in machine learning where multiple models, often referred to as "weak learners," are combined to solve a problem more effectively than any single model could on its own. This approach leverages the strengths and balances out the weaknesses of the individual models to improve overall performance, particularly in terms of accuracy and robustness.

Here's a brief introduction to the key ideas in ensemble learning:

1. **Basic Principle**:
   - The core idea behind ensemble learning is that a group of weak learners (models that are slightly better than random guessing) can come together to form a strong learner (a highly accurate model).
   - This is similar to how a group of people, each with their own individual expertise, can make better decisions collectively than a single expert.

2. **Types of Ensemble Methods**:
   - **Bagging (Bootstrap Aggregating)**: Involves building multiple models (usually of the same type) from different subsamples of the training dataset. An example is the Random Forest algorithm, which combines multiple decision trees with each tree trained on a random subset of the data.
   - **Boosting**: Focuses on training multiple models sequentially. Each new model gradually focuses more on the data points that previous models classified incorrectly. Examples include AdaBoost and Gradient Boosting.
   - **Stacking**: Involves training multiple different models and then using a new model to combine their predictions. The base level models are trained on the full training set, and then the combiner model is trained on the outputs of these base models.

Ensemble learning represents a significant step in the evolution of decision trees, transforming them from relatively simple and interpretable models to components of powerful, high-performance ensemble methods. In later classes, delving deeper into ensemble methods will reveal their potential for solving complex real-world problems that single models struggle with.
