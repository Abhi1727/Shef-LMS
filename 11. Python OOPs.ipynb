### OOPS (Object-Oriented Programming)

Imagine you're building with LEGOs. OOPS is like having different types of LEGO bricks (objects) that have specific shapes, colors (data), and can be connected in certain ways (methods). You can use these bricks to build various structures (programs).

### Components of OOPS

*   **Class:** Think of a blueprint for a house. It defines what a house will have (rooms, windows, doors) but isn't a physical house itself. In OOPS, a class is a blueprint for creating objects.
*   **Object:** A physical house built from the blueprint. You can have many houses (objects) from the same blueprint (class), and each house is a distinct entity.
*   **Variables:** The properties of your house, like the number of rooms, the color of the walls, or the type of roof. In OOPS, variables (attributes) store data about an object.
*   **Methods:** The actions you can perform with your house, like opening a door, turning on a light, or watering the garden. In OOPS, methods are functions within a class that define the behavior of an object.

### Class and Objects in Action

Your notebook demonstrates creating a `construct_house` class with a `design` attribute. This is like having a blueprint for houses that all have an "Urban" design. When you create `h1 = construct_house()`, you're building a specific house (object) based on that blueprint.

You then show how to access the `design` attribute of the `h1` object, just like you'd check the design of a physical house.

The example with `h1` shows creating another house with an "Arabic" design, 3 floors, and 5 rooms. This highlights that objects created from the same class can have different attribute values.

Adding the `add_pool` method to the `construct_house` class is like adding a feature to your house blueprint, allowing houses built from it to have a pool. When you call `h1.add_pool()`, you're using that feature on a specific house.

The calculator example with `simple_calci` demonstrates how a class can represent a concept with data (the numbers `a` and `b`) and methods (the `add` method) that operate on that data.

### Class Variable vs. Instance Variable

Imagine you have a set of building blocks (class).

*   **Class Variable:** A property that is the same for *all* the blocks in the set, like the brand name (e.g., LEGO). In your `construct_house` example, `design` and `floors` were initially treated as class variables, meaning all houses built from that blueprint had the same design and number of floors.
*   **Instance Variable:** A property that can be different for *each individual* block. If you paint some blocks red and some blue, the color is an instance variable. In your notebook, when you define `floors` and `rooms` within the `__init__` method using `self.floors` and `self.rooms`, you're making them instance variables. This allows each `construct_house` object (`h1` and `h2`) to have its own unique number of floors and rooms.

The `__init__` method is like the instruction manual for building a specific house. It tells you how to set up the initial features (instance variables) when you create a new house object.

The `calculator` example further illustrates instance variables (`a` and `b`) that are specific to each calculator object (`c1` and `c2`).

### Polymorphism

"Polymorphism" means "many forms." In OOPS, it means that you can have different objects that respond to the same method call in their own way.

Think of a "draw" action. If you tell a painter to "draw," they might draw a picture. If you tell a carpenter to "draw," they might draw a line on a piece of wood. The action "draw" has different forms depending on the object it's applied to.

Your `rectangle` and `triangle` classes with their own `area()` methods are a great example. Calling `area()` on a `rectangle` object calculates the area of a rectangle, while calling it on a `triangle` object calculates the area of a triangle. The `area()` method takes "many forms."



# Ploymorphism
The word "polymorphism" means "many forms", and in programming it refers to methods/functions/operators with the same name that can be executed on many objects or classes.

```python
''' create two class rectangle and triangle , implement area in both the classes'''
class rectangle:
  def __init__(self,l, b):
    self.l=l
    self.b=b
  def area(self):
    return self.l * self.b
class triangle:
  def __init__(self,base, h):
    self.base=base
    self.h=h
  def area(self):
    return 0.5*self.base *self.h #A=1/2*base*height
rec = rectangle(l=10,b=20)
tri = triangle(base=10,   h=5)
```

```python
# area of rectangle
rec.area()
```

```python
# area of triangle
tri.area()
```

## Inheritance

Inheritance is like passing down traits from parents to children. A child inherits characteristics from their parents.

*   **Single Inheritance:** A child inherits from one parent. Your `dog` class inheriting from the `Animal` class is an example. A dog is an animal, so it inherits the ability to `eat()` from the `Animal` class, and also has its own unique ability to `bark()`.
*   **Multiple Inheritance:** A child inherits from multiple parents. Your `son` class inheriting from both `Father` and `Mother` is an example. The son inherits the ability to `engineer()` from the `Father` class and `Singer()` from the `Mother` class, and also has his own ability to be a `student()`.
*   **Multilevel Inheritance:** A child inherits from a parent, and that parent inherited from a grandparent. Your `child` class inheriting from `father`, which inherited from `grand_father`, is an example. The child inherits traits from both their father and grandfather.
*   **Hierarchical Inheritance:** Multiple children inherit from a single parent. Your `Son` and `daughter` classes both inheriting from the `Father` class is an example. Both the son and daughter inherit the `father_name` from the `Father` class.

Your `simple_calci` and `scientific_calci` example is another illustration of single inheritance. The `scientific_calci` inherits the basic operations from `simple_calci` and adds its own scientific functions.

# Inheritance
Inheritance is a fundamental concept in Object-Oriented Programming (OOP) where a class (known as the child class or subclass) can inherit attributes and methods from another class (known as the parent class or superclass). This allows the child class to reuse code and extend the functionality of the parent class.

```python
class simple_calci:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def add(self):
    return self.a +self.b
  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a / self.b
  def sub(self):
    return self.a - self.b
class scientific_calci(simple_calci):
  def sq_root(self):
    return (self.a)**(0.5)
  def modulo(self):
    return self.a % self.b
simple = simple_calci(10,5)
scientific = scientific_calci(10,5)
```

```python
# operate simple calci
print(simple.a)
print(simple.b)
print(simple.add())
print(simple.mul())
print(simple.sub())
print(simple.div())
# print(simple.sq_root())
# print(simple.modulo())
# simple calci is parent class they have access only for 4 methods
```

```python
print(scientific.a)
print(scientific.b)
print(scientific.add())
print(scientific.mul())
print(scientific.sub())
print(scientific.div())
print(scientific.sq_root())
print(scientific.modulo())
```

### Encapsulation

Encapsulation is like putting related things into a box to keep them organized and protected. You decide what's inside the box and how others can interact with what's inside.

Think of a television. All the complex electronics and wiring are hidden inside the case (encapsulated). You interact with the TV through the remote control (methods) and don't need to touch the internal components directly.

Your `ATM` class demonstrates encapsulation. The `__balance` (the double underscore makes it a private variable, indicating it's not meant for direct external access) is the data (like the money in the ATM). The `check_balance()` method is the controlled way to access that data (like using the ATM interface to check your balance).

Initially, without the double underscore, you could directly change the `balance` attribute, which is like being able to magically change the amount of money in the ATM without using the proper interface. By making it `__balance` and providing a method `check_balance()` to access it, you are encapsulating the data and controlling how it's accessed, preventing unauthorized changes.

# Encapsulation
Encapsulation is the process of bundling data (attributes) and methods (functions) that operate on the data into a single unit, typically a class. It restricts direct access to some components, ensuring controlled interaction through methods.

* It prevents direct access to attributes.

# Abstraction
Abstraction means showing only the essential details of an object while hiding the unnecessary complexity. It allows you to focus on what something does rather than how it does it.

Think of driving a car:

* You use the steering wheel, accelerator, and brakes to drive.
* You don't need to know how the engine, gears, or braking system works internally.

Here, the car's controls are the abstracted interface that lets you drive without worrying about the technical details.

```python
from abc import ABC , abstractmethod
# abc abstract base class: Blue print we create for all the sub classes.
# abstractmethod: Defining method without implementation  in base class
# abc= library:   collection of all modules
# ABC = module: single .py file
```

```python
class ATM(ABC):# base class
  @abstractmethod
  def authenticate(self):
    pass
  @abstractmethod
  def check_balance(self):
    pass
  @abstractmethod
  def withdraw(self, w_amount):
    pass
  @abstractmethod
  def deposit(self, d_amount):
    pass
class simple_atm(ATM):
  def __init__(self, balance,authenticated):
    self.balance=balance
    self.authenticated=False
  def authenticate(self, card_no, pin):
    if card_no == 345678765 and pin == 1234:
      self.authenticated=True
      return "Verification is successful"
    else:
      return "Verification is failed"
  def check_balance(self):
    if self.authenticated==True:
      return (f"Current balance is {self.balance}")
  def withdraw(self, w_amount):
    if self.authenticated==True:
      if w_amount>self.balance:
        return "Your balance is not sufficient to withdraw"
      else:
        self.balance=self.balance-w_amount
        return f"Current balance after wihdrawing {w_amount} is {self.balance}"
  def deposit(self, d_amount):
    if self.authenticated==True:
      self.balance=self.balance + d_amount
      return f"current balance after depositing {d_amount} is {self.balance}"
obj= simple_atm(50000,False)
```

```python
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
  @abstractmethod
  def perimeter(self):
    pass
class rectangle(Shape):
  def __init__(self, l, b):
    self.l=l
    self.b=b
  def area(self):
    return (f"The area of the rectangle is {self.l*self.b}")
  def perimeter(self):
    return (f"The perimeter of the rectangle is {2*self.l + 2* self.b}")
  def get_details(self):
    return  (f"The length and breadth of the rectangle is {self.l} & {self.b}")
class circle(Shape):
  def __init__(self, radius):
    self.radius = radius
  def area(self):
    return 3.14*self.radius**2
  def perimeter(self):
    return round(2*3.14*self.radius,2)
  def get_details(self):
    return (f"The radius of the circle is {self.radius}")

objrect = rectangle(2,3)
objcircle = circle(10)
```

```python
print(objrect.get_details())
print(objrect.area())
print(objrect.perimeter())
```

```python
print(objcircle.area())
print(objcircle.perimeter())
print(objcircle.get_details())
```
