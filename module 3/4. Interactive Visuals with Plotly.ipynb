# Plotly: Interactive Data Visualization Library

## **What is Plotly?**
Plotly is a powerful Python library for creating interactive and aesthetically pleasing visualizations. It allows users to generate static, animated, and interactive plots for web-based or standalone applications.

Plotly supports:
- Integration with Python, R, JavaScript, MATLAB, and Julia.
- A wide range of visualization types, from simple scatter plots to complex 3D plots and dashboards.
- Interactive features like zooming, panning, and hover annotations.

---

## **Key Features of Plotly**
1. **Interactivity**:
   - Supports zoom, pan, and hover for data exploration.
   - Allows for drill-downs and detailed views.

2. **Wide Range of Visualizations**:
   - Basic charts: Line, Bar, Scatter, Pie.
   - Advanced charts: Tree Maps, Sunbursts, 3D Scatter Plots.
   - Statistical charts: Histograms, Box Plots, Distplots.
   - Business charts: Waterfall, Funnel, and Gauge Charts.
   - Heatmaps, Bubble Charts, and more.

3. **Customization**:
   - Highly customizable layouts, colors, and markers.
   - Responsive designs for various screen sizes.

4. **Integration**:
   - Seamlessly integrates with Jupyter Notebooks and Dash for web-based applications.
   - Works well with pandas and NumPy for data handling.

5. **Web-based Sharing**:
   - Visualizations can be saved as HTML files or embedded into web applications.
   - Easy sharing through static or interactive formats.

6. **Advanced Analytics**:
   - Built-in support for statistical and financial charts.
   - Capable of creating 3D plots for in-depth visualizations.

---

## **Why Use Plotly?**
1. **User-Friendly Interactivity**:
   - Unlike static libraries like Matplotlib or Seaborn, Plotly provides an interactive experience, enabling deeper insights through direct user engagement.

2. **Dynamic Dashboards**:
   - Perfect for real-time monitoring or presentation of Key Performance Indicators (KPIs) in dashboards.

3. **Customization & Professional Aesthetics**:
   - Enhances the visual appeal and provides better control over chart appearance compared to traditional libraries.

4. **Web-Ready Visualizations**:
   - Easily share plots online or embed them in websites and applications.

5. **Broad Application Range**:
   - Suitable for exploratory data analysis, business intelligence, reporting, and educational purposes.

---

## **What Can Be Done with Plotly?**
1. **Basic Charts**:
   - Line, Bar, Scatter, Pie, and Histogram.

2. **Statistical Charts**:
   - Box Plots, Violin Plots, Distplots.

3. **3D Visualizations**:
   - 3D Scatter, 3D Surface, and 3D Mesh Plots.

4. **Business Analytics**:
   - Waterfall, Funnel, and Gauge Charts.

5. **Hierarchical Charts**:
   - Tree Maps, Sunburst Charts, and Packed Circle Charts.

6. **Heatmaps**:
   - Visualize correlations or intensities.

7. **Custom Dashboards**:
   - Create interactive dashboards with `Dash`, a Python framework built on Plotly.

---

## **Advantages of Plotly**
- Interactive and dynamic features for a superior user experience.
- Advanced visualization options unavailable in libraries like Matplotlib or Seaborn.
- Ideal for real-time data dashboards and presentations.
- Great for web applications with easy HTML export and embedding.

---

Plotly is an essential tool for anyone looking to enhance their data visualization capabilities, whether for business insights, academic research, or professional presentations.

# 📈 Line Chart

## **What**
A line chart is used to visualize trends over time or continuous data.

## **Why**
- Analyze changes over time.
- Identify trends, patterns, or anomalies.
- Useful for time-series data like sales, stock prices, or weather.

## **Syntax**
```python
import plotly.express as px

fig = px.line(data_frame, x="time_column", y="numeric_column", title="📈 Line Chart")
fig.show()


**What is the yearly trend in the number of YouTube channels created?**

````python
data['created_year'].value_counts().reset_index().sort_values(by='created_year',ignore_index=True)
````

````python
# Count the number of channels created per year
channels_per_year = data['created_year'].value_counts().reset_index()
channels_per_year = channels_per_year.sort_values(by='created_year')
````

````python
fig_subscriber_growth = px.line(
    channels_per_year,
    x='created_year', y='count',
    title="Line Chart: Subscriber Growth Over the Years"
)
fig_subscriber_growth.show()
````

**Insights for Subscriber Growth Over the Years**
* **Consistent Growth:** Subscriber counts grew steadily from 1970 to 2000, reflecting a gradual adoption of content.
* **Rapid Expansion:** After 2000, there was a sharp increase in subscriber counts, peaking around 2010.
* **Recent Decline:** Subscriber growth has decreased after 2010, possibly due to market saturation or shifting content trends.
* **Historical Trends:** The chart highlights major milestones in subscriber adoption over decades, aligning with advancements in digital platforms.

* **Steady Increase:** Subscriber growth was steady from 1970 to 2000, indicating gradual adoption.

# ✨ Scatter Plot

## **What**
A scatter plot shows the relationship between two numerical variables.

## **Why**
- Identify correlations or trends between two variables.
- Visualize clusters or patterns in the data.
- Useful for regression analysis and anomaly detection.

## **Syntax**
```python
import plotly.express as px

**Is there a relationship between engagement rate and total video views across channels?**



````python
fig_engagement_vs_views = px.scatter(
    data, x='video views', y='engagement_rate',
    title="Scatter Plot: Engagement Rate vs Total Video Views",
    size='subscribers', color='Country',
    hover_name='Youtuber',
    labels={'video views': 'Total Video Views', 'engagement_rate': 'Engagement Rate'}
)
fig_engagement_vs_views.show()
````

* **Clusters:** The majority of countries fall within the lower engagement and view range, reflecting typical audience behavior.
* **Opportunity Areas:** Countries with mid-level views and engagement could be untapped for optimization.

````python
import plotly.figure_factory as ff

# Assuming 'data' is your dataframe
correlation = data[['subscribers', 'video views', 'engagement_rate', 'uploads']].corr()

fig_heatmap = ff.create_annotated_heatmap(
    z=correlation.values,
    y=list(correlation.columns),
    x=list(correlation.index),
    colorscale='Viridis'
)

# Adding title to the layout
fig_heatmap.update_layout(
    title="Heatmap: Correlation Between Numeric Variables",
    xaxis_title="Variables",
    yaxis_title="Variables"
)

fig_heatmap.show()
````

**Insights for Heatmap: Correlation Between Numeric Variables**
* **Strong Positive Correlation:** Subscribers and Video Views have a high correlation (0.85), indicating that channels with more subscribers generally have higher video views.

# **What is a Waterfall Chart?**
A waterfall chart is a visualization tool used to display the cumulative effect of sequentially introduced positive or negative values. It is commonly used in financial and operational analysis to show how an initial value is influenced by a series of intermediate positive and negative changes, leading to a final value.

**Uses of a Waterfall Chart:**

* Understanding financial data, such as revenue breakdown or profit analysis.
* Visualizing changes in metrics over time or across categories.
*Explaining contributions to a total, such as product sales, operating expenses, etc.


General syntax


```
import plotly.graph_objects as go

fig = go.Figure(
    go.Waterfall(
        x=["Category A", "Category B", "Category C", "Total"],  # Categories
        y=[100, -20, 30, 110],  # Contributions (positive/negative)
        connector={"line": {"color": "gray"}}  # Style of connectors
    )
)

fig.update_layout(
    title="Waterfall Chart Example",
    xaxis_title="Categories",
    yaxis_title="Values"
)

fig.show()
```

# What is a Subplot?
A **subplot** is a way to display multiple plots within a single figure, arranged in a grid structure. Each plot occupies a specific position in the grid, allowing for easy comparison of multiple visualizations.

In **Plotly**, subplots are created using the `make_subplots()` function, where you can define the number of rows, columns, and customization for each subplot.

---

### Keywords in the Code

#### **1. `make_subplots()`**
- **Purpose**: Creates a grid layout for multiple plots.
- **Arguments**:
  - `rows` and `cols`: Define the grid structure (e.g., `rows=2, cols=2` for a 2x2 grid).
  - `subplot_titles`: Adds titles above each subplot.

---

#### **2. Plot Types**
- **`go.Histogram`**: Creates a histogram for frequency distribution.
  - `x`: Data for the histogram (e.g., `data['subscribers']`).
  - `name`: The trace name (used in the legend).
  
- **`go.Scatter`**: Creates scatter or line plots.
  - `x`: Data for the x-axis (e.g., `data['rank']`).
  - `y`: Data for the y-axis (e.g., `data['video views']`).
  - `mode`: Specifies the style of the plot (`lines+markers` or `markers`).
  - `name`: The trace name (used in the legend).
  
- **`go.Bar`**: Creates a bar chart for comparison.
  - `x`: Categories for the bars (e.g., `engagement_data.index`).
  - `y`: Values for each bar (e.g., `engagement_data.values`).
  - `name`: The trace name (used in the legend).

---

#### **3. `fig.add_trace()`**
- **Purpose**: Adds a plot (trace) to a specific subplot.
- **Arguments**:
  - Plot type (`go.Histogram`, `go.Scatter`, etc.).
  - `row` and `col`: Position of the plot in the grid (e.g., `row=1, col=1` for the first subplot).

---

#### **4. `fig.update_xaxes()` and `fig.update_yaxes()`**
- **Purpose**: Customize the x-axis and y-axis for a specific subplot.
- **Arguments**:
  - `title_text`: Axis title (e.g., "Subscribers").
  - `row` and `col`: Subplot location to apply changes.

---

#### **5. `fig.update_layout()`**
- **Purpose**: Configures the overall layout of the figure.
- **Arguments**:
  - `height` and `width`: Dimensions of the figure.
  - `title_text`: Main title for the entire figure.
  - `showlegend`: Shows or hides the legend.

---

#### **6. `groupby()`**
- **Purpose**: Groups data by a specific column and applies aggregation (like mean or sum).
- **Example**:
  ```python
  engagement_data = data.groupby('category')['engagement_rate'].mean()


````python
# Import necessary modules from Plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a subplot layout with 2 rows and 2 columns
# Add custom titles for each subplot
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Histogram of Subscribers",  # Top-left: Histogram
                    "Line Plot: Video Views vs Rank",  # Top-right: Line Plot
                    "Bar Chart: Engagement Rate by Category",  # Bottom-left: Bar Chart
                    "Scatter Plot: Subscribers vs Video Views")  # Bottom-right: Scatter Plot
)

# Add a Histogram to the first subplot (row 1, col 1)
fig.add_trace(
    go.Histogram(x=data['subscribers'], name="Subscribers Histogram"),  # Plot subscribers as a histogram
    row=1, col=1  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the histogram
fig.update_xaxes(title_text="Subscribers", row=1, col=1)
fig.update_yaxes(title_text="Frequency", row=1, col=1)

# Add a Line Plot to the second subplot (row 1, col 2)
fig.add_trace(
    go.Scatter(x=data['rank'], y=data['video views'], mode='lines+markers', name="Video Views vs Rank"),  # Line plot with markers
    row=1, col=2  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the line plot
fig.update_xaxes(title_text="Rank", row=1, col=2)
fig.update_yaxes(title_text="Video Views", row=1, col=2)

# Prepare data for the Bar Chart: Calculate average engagement rate by category
engagement_data = data.groupby('category')['engagement_rate'].mean().sort_values(ascending=False)

# Add a Bar Chart to the third subplot (row 2, col 1)
fig.add_trace(
    go.Bar(x=engagement_data.index, y=engagement_data.values, name="Engagement Rate by Category"),  # Bar chart with categories and average engagement rates
    row=2, col=1  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the bar chart
fig.update_xaxes(title_text="Category", row=2, col=1)
fig.update_yaxes(title_text="Engagement Rate", row=2, col=1)

# Add a Scatter Plot to the fourth subplot (row 2, col 2)
fig.add_trace(
    go.Scatter(x=data['subscribers'], y=data['video views'], mode='markers',  # Scatter plot with subscribers and video views
               name="Subscribers vs Video Views"),
    row=2, col=2  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the scatter plot
fig.update_xaxes(title_text="Subscribers", row=2, col=2)
fig.update_yaxes(title_text="Video Views", row=2, col=2)

# Update the overall layout of the figure
fig.update_layout(
    height=800,  # Set the height of the entire figure
    width=1000,  # Set the width of the entire figure
    title_text="Subplot with 4 Different Graphs",  # Title of the figure
    showlegend=True  # Enable the legend to show plot labels
)

# The plot can now be displayed using fig.show() or saved as an HTML file
````

# **What is Interactivity?**
Interactivity in data visualization refers to the ability to interact dynamically with a chart or graph to explore, filter, or drill down into data. Features like dropdown menus, zooming, panning, and tooltips are examples of interactivity. Plotly provides interactive elements that let users:

* Filter Data: Use dropdowns or sliders to focus on specific subsets of data.
* Hover: See detailed information by hovering over data points.
* Zoom/Pan: Explore specific regions of a plot dynamically.

````python
data.groupby(['category', 'Country'])['video views'].sum().reset_index()
````

````python


# Aggregate data by category and country to sum video views
category_country_views = data.groupby(['category', 'Country'])['video views'].sum().reset_index()

# Create the initial bar chart
fig = px.bar(
    category_country_views,
    x='category',
    y='video views',
    color='category',
    title="Total Video Views by Category",
    labels={'video views': 'Video Views', 'category': 'Category'}
)

# Add dropdown for countries (multi-select logic implemented manually)
unique_countries = category_country_views['Country'].unique()

# Create a dropdown for country selection
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label="All Countries",
                     method="update",
                     args=[{"visible": [True] * len(category_country_views)},
                           {"title": "Total Video Views by Category (All Countries)"}]),
                *[
                    dict(label=country,
                         method="update",
                         args=[{"visible": category_country_views['Country'] == country},
                               {"title": f"Total Video Views by Category ({country})"}])
                    for country in unique_countries
                ],
                dict(label="Clear",
                     method="update",
                     args=[{"visible": [False] * len(category_country_views)},
                           {"title": "Total Video Views by Category (No Country Selected)"}])
            ],
            direction="down",
            showactive=True,
            x=0.1,
            y=1.15,
        )
    ]
)

# Show the chart
fig.show()
````

# 📈 **Best Practices and Tips for Plotly**

## 🌟 **Best Practices**
1. **Start with `plotly.express`:**
   - Use `plotly.express` for quick visualizations and default aesthetics. Switch to `plotly.graph_objects` for advanced customization.

6. **Integrate Dash for Complex Interactions:**

7. **Choose the Right Chart Type:**
