# **What is an API?** 🌐🔗
An API (Application Programming Interface) is a toolbox 🛠️ that allows different software applications to communicate and share data with each other. It’s like a bridge 🌉 that connects two systems, enabling them to interact without needing to know how the other works internally.

image.png

## **Key Characteristics of APIs:** 📋
1. **Intermediary**: APIs act as a middleman 🤝 between systems to ensure smooth communication.
2. **Abstraction**: They hide the complex details 🧩 of the system, providing only what is necessary for developers.
3. **Standardization**: APIs follow consistent rules 🗂️ (e.g., HTTP methods) for easy communication.

# **How to Use APIs and Perform CRUD Operations**
The script demonstrates how to interact with a **RESTful API** to perform **CRUD operations** on an employee database. To understand how it works, let’s break it down into the key concepts of RESTful APIs, HTTP protocols, and the use of Python’s requests library.

## **What is a RESTful API?**
API: Stands for Application Programming Interface. It allows applications to communicate with each other by exchanging data.

RESTful API: Follows the principles of REST (Representational State Transfer), a widely-used architecture for building APIs.


### **Key Features of RESTful APIs:**

1. **Resource-based:**
Resources (like employees, customers, products) are represented by URLs (e.g., /employees, /employees/{id}).

2. **Stateless:**
Each API request contains all the information needed for the server to process it; no client data is stored on the server between requests.

3. **HTTP Protocol:**
RESTful APIs leverage HTTP methods like GET, POST, PUT, PATCH, and DELETE for CRUD operations.

4. **JSON Format:**
Data is typically exchanged in JSON format, which is lightweight and easy to parse.

### **HTTP Methods and CRUD Operations**
**HTTP Methods:**

1. **GET:** Retrieves data from the server.
2. **POST:** Sends data to the server to create a new resource.
3. **PUT:** Updates or replaces an entire resource.
4. **PATCH:** Updates specific fields of a resource.
5. **DELETE:** Deletes a resource from the server.

## **How Python's `requests` Library Works**

The `requests` library in Python simplifies making HTTP requests. It provides methods to send requests and handle responses.

### **Basic Workflow**

#### **1. Send a Request**
- Use methods like `get()`, `post()`, `put()`, `patch()`, or `delete()` to send an HTTP request to the server.
- Specify the **endpoint (URL)** and include additional data (like JSON payloads for `POST` or `PUT`, if needed).

#### **2. Receive a Response**
- The server responds with:
  - A **status code** indicating the outcome of the request.
  - Optional **data** (usually in JSON format).
- Use `response.json()` to parse the JSON data.

#### **3. Handle Status Codes**

| **Status Code**       | **Description**                                     |
|------------------------|-----------------------------------------------------|
| **200 OK**             | The request was successful.                        |
| **201 Created**        | A new resource was successfully created.           |
| **404 Not Found**      | The requested resource doesn’t exist.              |
| **400 Bad Request**    | The request was invalid (e.g., missing fields).    |
| **500 Internal Server Error** | The server encountered an error.             |

## **What is the BASE_URL?**
The BASE_URL is the root address of the API server. All requests are directed to this base URL, with specific endpoints appended to it for performing various operations.

**In this case:**

BASE_URL = http://98.70.13.15:8787

This URL serves as the entry point to the API.

**For example:**
1. To retrieve all employees: http://98.70.13.15:8787/employees
2. To retrieve a specific employee by ID: http://98.70.13.15:8787/employees/{Specific_id}
3. To add a new employee: http://98.70.13.15:8787/employees (via POST method)

The BASE_URL provides flexibility by centralizing the API's address. If the server's location changes, updating the BASE_URL ensures all requests are redirected to the new address without modifying each function.

### **What is an Endpoint?**

In the context of APIs (Application Programming Interfaces), an endpoint is a specific URL where a particular resource or functionality of the API can be accessed. It is the destination for an API request and defines where the server should listen for requests from a client.

**Components of an Endpoint**

An endpoint is typically composed of:

1. Base URL: The root address of the API.

Example: http://api.example.com

2. Path: Specifies the resource or operation you want to perform.

Example: /employees, /employees/{id}

**When combined, they form a complete URL:**

http://api.example.com/employees

## **2. POST** 📤
- **Purpose:** Send data to the server to create a new resource.
- **Use Case:** When you need to add new information, such as creating a new user, uploading a photo, or submitting a form.
- **Example:**
  - Request: `POST /users`
  - Payload: `{"name": "Alice", "age": 25}`
  - Response: Confirmation of the new user creation ✅.
- **Analogy:** Like mailing a package 📦—you’re sending something to be added.

# **Test POST to add a new employee**

````python
def test_add_employee():
    new_employee = {
        "id": 10,
        "name": "emma",
        "age": 30,
        "gender": "female",
        "department": "sales",
        "salary": 900000
    }
    response = requests.post(f"{BASE_URL}/posts", json=new_employee)
    print("POST /employees:", response.json())
    print("Status Code", response.status_code)
````

````python
#Adds a new employee to the database
test_add_employee()
````

This function adds a new employee to the database. It defines a dictionary with the employee's details, such as `id`, `name`, `age`, `gender`, and `salary`. A `POST` request is sent to the `/employees` endpoint with this data as a JSON payload. The server responds with a confirmation and the details of the new employee, which the function prints.

## **3. PUT** 📝
- **Purpose:** Update an existing resource completely.
- **Use Case:** When you need to modify an entire resource, such as updating a profile or replacing data.
- **Example:**
  - Request: `PUT /users/123`
  - Payload: `{"name": "Alice", "age": 26}`
  - Response: The updated user data 🔄.
- **Analogy:** Like editing a whole document 📄.

# **Test PUT to update an employee (replace completely or add)**

````python
def test_update_employee(emp_id):
    updated_employee = {
        "id": emp_id,  # ID must match the one being updated
        "name": "emma_updated",
        "age": 31,
        "gender": "female",
        "department": "sales",
        "salary": 950000
    }
    response = requests.put(f"{BASE_URL}/posts/{emp_id}", json=updated_employee)
    print(f"PUT /employees/{emp_id}:", response.json())
````

````python
# Fully updates the record of the employee with ID 10.
test_update_employee(10)
````

This function replaces the entire record of an employee with updated data. It creates a dictionary containing the new details for the employee and sends a `PUT` request to the `/employees/{id}` endpoint, where `{id}` is the employee's unique ID. The server updates the record and returns the updated details, which the function prints.

## **4. PATCH** 🩹
- **Purpose:** Partially update an existing resource.
- **Use Case:** When you need to modify only specific fields, such as changing just the username without altering other details.
- **Example:**
  - Request: `PATCH /users/123`
  - Payload: `{"name": "Alice B."}`
  - Response: Updated user data 🎯.
- **Analogy:** Like fixing a typo(typing mistake) ✍️ in a document.

# **Test PATCH to partially update an employee**

````python
def test_patch_employee(emp_id):
    updates = {
        "age": 32,
        "salary": 1000000
    }
    response = requests.patch(f"{BASE_URL}/employees/{emp_id}", json=updates)
    print(f"PATCH /employees/{emp_id}:", response.json())
````

````python
# Partially Update Employee
test_patch_employee(10) # Partially update employee with ID 10
````

This function partially updates an employee's record. It defines a dictionary containing only the fields to be updated, such as `age` or `salary`. A `PATCH` request is sent to the `/employees/{id}` endpoint, and the server applies the changes to the record. The server responds with the updated details, which the function prints.

## **5. DELETE** 🗑️
- **Purpose:** Remove a resource from the server.
- **Use Case:** When you need to delete information, such as removing a user or deleting a file.
- **Example:**
  - Request: `DELETE /users/123`
  - Response: Confirmation of deletion 🛑.
- **Analogy:** Like throwing something away 🗑️—it’s gone.

# **Test DELETE to remove an employee**

This function removes an employee from the database by sending a `DELETE` request to the `/employees/{id}` endpoint, where `{id}` is the employee's unique ID. The server deletes the record and returns a confirmation message, which the function prints for verification.
