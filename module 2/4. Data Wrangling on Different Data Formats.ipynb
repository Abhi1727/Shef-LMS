### **📝Understanding Different Data Formats and Their Techniques for Data Wrangling**

image.png


**📊 Structured Data:** This type of data is organized in a tabular format, with rows and columns. It's typically stored in relational databases, spreadsheets, or CSV (comma-separated values) files. Data wrangling on structured data involves cleaning, transforming, and aggregating data to make it suitable for analysis.


Some common techniques used for data wrangling on structured data include:


**🧼 Data cleaning:** removing duplicates, filling in missing values, and correcting errors. 🔄 Data transformation: converting data types, renaming columns, and splitting or merging columns. 🔍 Data aggregation: summarizing data, grouping data by categories, and calculating statistics.


**🔀 Unstructured Data:** This type of data does not have a predefined structure, such as text, images, audio, and video. Data wrangling on unstructured data involves extracting relevant information from the data and transforming it into a structured format.




**Some common techniques used for data wrangling on unstructured data include:**


**📚 Text mining:** extracting information from unstructured text data using techniques such as natural language processing and sentiment analysis. 📷 Image and video processing: extracting features and patterns from images and videos using techniques such as object detection and facial recognition. 🎤 Audio processing: analyzing audio data to extract features such as pitch, volume, and speech recognition.


In summary, structured data is organized in a tabular format, while unstructured data doesn't have a predefined structure. Data wrangling techniques for structured data include data cleaning, data transformation, and data aggregation, while techniques for unstructured data include text mining, image and video processing, and audio processing. These techniques help to extract relevant information from the data and transform it into a structured format, making it suitable for analysis.

##<b>6️⃣ Date and Time libraries

dateimeg.jpg

📅 The datetime library is like a data timekeeper!💎.  

 The datetime library provides functions for working with dates and times. It allows you to create, manipulate, and format dates and times in a variety of ways. With this library, you can perform operations such as calculating time differences, converting between time zones, and parsing dates from text strings.


The datetime module in Python provides classes for working with dates and times. It allows you to create and manipulate dates, times, and timedeltas (differences between dates and times).

Here's an example of how to use datetime in Python: 👇

### **Get todays date and extract date components(day,month,year,dayname,monthname)**

````python
import datetime

# Get the current date and time
now = datetime.datetime.now()

# Print the current date and time
print(now)
print(type(now))
````

````python
# Extract date components
day = now.day  # Day of the month (1–31)
month = now.month  # Month as a number (1–12)
year = now.year  # Year
day_name = now.strftime("%a")  # Full name of the day (e.g., "Monday")
month_name = now.strftime("%b")  # Full name of the month (e.g., "June")

# Print the extracted components
print(f"Current Date and Time: {now}")
print(f"Day of the Month: {day}")
print(f"Month (Number): {month}")
print(f"Year: {year}")
print(f"Day Name: {day_name}")
print(f"Month Name: {month_name}")
````

* now.day: Extracts the day of the month (1–31).
* now.month: Extracts the month as a number (1–12).
* now.year: Extracts the year (e.g., 2024).
* strftime("%A"): Returns the full name of the day (e.g., "Monday").
* strftime("%B"): Returns the full name of the month (e.g., "June").
* strftime("%a"): Returns the short name of the day (e.g., "Mon").
* strftime("%b"): Returns the short name of the month (e.g., "Jun").

1. %A → Full day name (e.g., Monday).
2. %a → Short day name (e.g., Mon).
3. %B → Full month name (e.g., June).
4. %b → Short month name (e.g., Jun).
5. %d → Day of the month (e.g., 17).
6. %m → Month number (e.g., 06).
7. %Y → Year (e.g., 2024).


### **Extract time components(hour,minute,second)**

````python

# Extract time components
hour = now.hour         # Hour (0–23)
minute = now.minute     # Minute (0–59)
second = now.second     # Second (0–59)

# Print the extracted components
print(f"Current Time: {now}")
print(f"Hour: {hour}")
print(f"Minute: {minute}")
print(f"Second: {second}")
````

**Adding and subtracting days**

### **Timedelta:**

The **timedelta** class in Python's datetime module represents the difference between two dates or times. It is used to perform operations like addition or subtraction with dates and times.

**timedelta:**

Use it for days, hours, minutes, and seconds.

````python
from datetime import datetime, timedelta

# Current date and time
now = datetime.now()
print("Now:", now)

# Add 2 days
future = now + timedelta(days=2)
print("Future:", future)

# Subtract 3 hours
past = now - timedelta(hours=3)
print("Past:", past)
````

### **Relativedelta**
The relativedelta class from the dateutil module is designed to handle operations involving years, months, days, hours, and more. Unlike timedelta, it can accurately handle months and years, which vary in length.

````python
from datetime import datetime
from dateutil.relativedelta import relativedelta

# Current date
now = datetime.now()
print("Current Date:", now)

# Add 2 months
future_2_months = now + relativedelta(months=2)
print("After 2 Months:", future_2_months)

# Subtract 2 months
past_2_months = now - relativedelta(months=2)
print("Before 2 Months:", past_2_months)

# After 2 years
after_2_years = now + relativedelta(years=2)
print("After 2 Years:", after_2_years)

# After 3 years
after_3_years = now + relativedelta(years=3)
print("After 3 Years:", after_3_years)

# Before 2 years
before_2_years = now - relativedelta(years=2)
print("Before 2 Years:", before_2_years)

# Before 3 years
before_3_years = now - relativedelta(years=3)
print("Before 3 Years:", before_3_years)
````

* relativedelta(years=2): Adds 2 years to the current date.
*   -relativedelta(years=3): Subtracts 3 years from the current date.

Works similarly for months, days, hours, etc., by replacing years with months, days, etc.

### **Converting String to Datetime**
In Python, you can convert a string to a date/time/datetime object using the datetime module's **strptime()** function.

 The **strptime()** function takes two arguments: the string to convert and a format string that specifies the format of the date/time/datetime string.

For example, to convert a string "2023-04-12" to a date object, you can use the following code:

````python
from datetime import datetime

date_string = '2023-04-12'
print(type(date_string))
date_object = datetime.strptime(date_string, '%Y-%m-%d').date()

print(date_object)
print(type(date_object))
````

This code first imports the datetime module and defines the date string to convert. It then uses the **strptime()** function to convert the date string to a date object. The format string '%Y-%m-%d' specifies that the date string has the year first, followed by the month and day separated by hyphens.

### **Converting Datetime to String**
Similarly, to convert a datetime object to a string, you can use the **strftime()** function. The **strftime()** function takes a format string as an argument and returns a string that represents the datetime object formatted according to the format string.

For example, to convert a datetime object to a string in the format "2023-04-12 12:30:45", you can use the following code:

````python
from datetime import datetime
datetime_object = datetime(2023, 4, 12, 12, 30, 45)
datetime_string = datetime_object.strftime('%Y-%m-%d %H:%M:%S')
print(datetime_string)
print(type(datetime_string))
````

This code first defines a datetime object with the year, month, day, hour, minute, and second. It then uses the strftime() function to format the datetime object as a string in the format '%Y-%m-%d %H:%M:%S', which specifies that the year comes first, followed by the month, day, hour, minute, and second separated by colons.

**Points to remember**
* strptime: Converts a string to a datetime object using a specified format.
* strftime: Converts a datetime object to a string using a specified format.
* Format codes like %Y, %m, %d, %H, and %M are essential for interpreting or formatting dates.
