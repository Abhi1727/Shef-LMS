# **What is Data Wrangling??**
Data wrangling, also known as data cleaning or data munging, is the process of transforming raw data into a clean and organized format, ready for analysis. This involves identifying and addressing issues within the dataset to make it suitable and reliable for further insights or machine learning models.

**Key Steps in Data Wrangling**

	1.	Data Cleaning: Removing or handling missing values, correcting inconsistencies, and dealing with outliers.
	2.	Data Structuring: Converting data into a structured format, often by reshaping or reformatting columns and rows.
	3.	Data Transformation: Changing data types, creating new calculated columns, or encoding categorical variables.
	4.	Data Filtering and Selection: Selecting specific subsets of data based on conditions or criteria.
	5.	Data Merging: Combining multiple datasets into a single, cohesive dataset.

**Why Data Wrangling is Important**??

	•	Ensures Data Quality: Improves the accuracy and reliability of data, which leads to more accurate analysis and insights.
	•	Prepares for Analysis: Makes data easier to explore, analyze, and visualize by organizing it in a consistent format.
	•	Increases Efficiency: Reduces time and effort required for analysis by resolving issues in the data early on.

Data wrangling is a critical step in the data science workflow, ensuring that data is clean, organized, and ready for meaningful analysis.

# **Indexing and Slicing**
🔍 Indexing refers to the process of accessing a single item in a sequence by its position, also known as its index. In Python, indexing starts at 0, so the first item in a sequence has an index of 0, the second item has an index of 1, and so on. For example, given the list a = [1, 2, 3, 4], you can access the value of the first item using a[0].

🔪 Slicing refers to the process of accessing multiple items in a sequence by specifying a range of indices. In slicing, you specify the start and end indices separated by a colon (:), and the slicing operation returns a new sequence that contains all the items in the original sequence that have an index between the start and end indices. For example, given the list a = [1, 2, 3, 4], you can extract the sublist [2, 3] using the slice a[1:3].

Both indexing and slicing are essential tools for working with large datasets in Python, allowing you to extract specific pieces of information quickly and efficiently. So go ahead, slice and dice your data to your heart's content! 🎉

###**syntax**

*  by: The column(s) to sort by.
* ascending: Sort order; True for ascending (default) and False for descending.

When using **ignore_index=True** in the sort_values() method in pandas, it ensures that the resulting DataFrame has its index reset to a default integer index after sorting. This is particularly useful when you no longer need the original index after sorting.

**ignore_index:**
* When combining, sorting, or resetting the index in a DataFrame, setting **ignore_index=True** reassigns a new integer index to the result, starting from 0, and discards the existing index.


**inplace=True**:

* inplace=True in Pandas is a parameter that allows you to perform an operation directly on the original DataFrame, modifying it “in place” without creating or returning a new copy of the DataFrame. This can be useful for saving memory and keeping the code concise.
* When inplace=True is used, the original DataFrame is altered, and the method does not return anything (it returns None).

**ascending=False in Pandas**
In Pandas, the ascending parameter is used in sorting functions like sort_values() to specify whether the data should be sorted in ascending or descending order.

* ascending=True: Sorts the data in ascending order (default behavior).
* ascending=False: Sorts the data in descending order.

# **Filter data using boolean indexing**
Filtering data using boolean indexing is a method in data analysis that isolates rows in a dataset based on specified conditions. In this process, conditions are applied to columns within a DataFrame, resulting in a boolean Series (a series of True/False values) that reflects whether each row meets the criteria. This boolean Series is then used to index the DataFrame, returning only the rows where the condition evaluates to True. Boolean indexing is a fast and efficient way to filter data, allowing for simple conditions (like exact matches) as well as more complex conditions using logical operators (such as & for “and” or | for “or”) without the need for loops or manual row selection.

### **Filter data using single condition**
Use comparison operations to filter data based on single condition

**data.loc[df['column_name'] == value]**

```python
# find all the movies released in english
data.loc[data.loc[:,'orig_lang']=='english']
```

```python
#  find all the movies released in hindi
hindi_movies = data.loc[data.loc[:,'orig_lang']=='hindi']
hindi_movies
```

```python
# find all the movie released after 2014
data.loc[data.loc[:,'year']>2014]
```

```python
# find all the movie which has given negative profits
data.loc[data.loc[:,'profit']<0]
```

# **Filter data using multiple conditions**
To filter data based on multiple conditions we use logical operators like and(&) , or(|) and not(!=).

**data.loc[(data['column']== condition) operator (data['column']== condition) ]**

```python
# find all the english movies released in 2012
data.loc[(data.loc[:,'orig_lang']=='english') & (data.loc[:,'year']==2012)]
```

```python
# find all the hindi movies giving negative profit
data.loc[(data.loc[:,'orig_lang']=='hindi') & (data.loc[:,'profit']<0)]
```

```python
# find all movies released in either 2012 or 2013
data.loc[(data.loc[:,'year']==2012) | (data.loc[:,'year']==2013)]
```

```python
# find all movies released in either english or hindi
data.loc[(data.loc[:,'orig_lang']=='hindi') |(data.loc[:,'orig_lang']=='english')]
```
