## 🚀 **What is the ML Lifecycle?**




The **Machine Learning Lifecycle** is the step-by-step process involved in building an ML model — from understanding the problem to deploying a working solution and monitoring it.

### 🌿 **Think of it like growing a plant:**

- **You start with the right seed**  
  *(problem & data)*

- **You nurture it**  
  *(clean and train)*

- **You make sure it grows well**  
  *(evaluate & tune)*

- **And finally, you show it to the world 🌍**  
  *(deploy & monitor)*

Each stage is crucial to ensure your model is accurate, reliable, and useful in real-world scenarios.

### 🎯 **Why is it Important?**



Without following a proper lifecycle:

- Models may **fail to generalize** 🤕  
- Insights may be **wrong** ❌  
- Businesses may **lose money** 💸  

✅ So, **understanding the ML lifecycle** ensures a **repeatable**, **scalable**, and **successful** ML project.

###**Data Ingestion | Data Collection**


#### **What is Data ingestion?**


**Data ingestion** is the first and foundational step of any machine learning pipeline.  
It refers to the process of **loading raw data** from various sources into the system for further processing.

🔧 **Purpose of the Component**

- Connects to the **data source**
- Reads the **raw dataset** (e.g., CSV, JSON, database, API)
- Stores it in a usable format (e.g., Pandas DataFrame)
- Passes it to the next stage of the ML lifecycle (Data Cleaning / EDA)



#### **What is Data Collection?**


Data Collection refers to the process of gathering raw data from various sources (like databases, APIs, web scraping, files, sensors, etc.) which will be used to train, validate, or test a machine learning model.

✅ **Purpose:**

To gather relevant, diverse, and high-quality data that reflects the real-world problem you're trying to solve with ML.

#### **Data Collection vs Data Ingestion**

| Term                | When to Use                                   | Meaning                                                                 |
|---------------------|-----------------------------------------------|-------------------------------------------------------------------------|
| 📊 **Data Collection** | When you're **gathering data from the real world** | This includes **scraping websites**, **calling APIs**, **querying databases**, **extracting sensor data**, or **collecting user activity logs**. |
| 📥 **Data Ingestion**  | When you're **loading already collected data** into your pipeline or system | This involves **reading a CSV, JSON, or database table** into your code (usually into a **DataFrame**). |

### 🧼 **Data Preprocessing / Cleaning**

#### 🔧 **Key Steps:**



1. **Handle Missing Values**
- Fill with **mean**, **median**, **mode**, or **drop** rows/columns

2. **Remove Duplicates**
- Eliminate repeated rows

3. **Handle Outliers**
- Detect using **IQR**, **Z-score**, or **visualization**
- Treat by **capping**, **transformation**, or **removal**

4. **Encode Categorical Variables**
- **Label Encoding** for ordered categories
- **One-Hot Encoding** for unordered categories

5. **Scale Numerical Features**
- **Normalize** (0–1 range)  
- **Standardize** (mean = 0, std = 1)

6. **Convert Data Types**
- Ensure correct types: `int`, `float`, `datetime`, `category`

````python
# preprocessing.py
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
import pickle

def preprocess_and_split(path='data/heart.csv'):
    # 📥 Load the dataset
    df = pd.read_csv(path)
    print(f"✅ Loaded data with shape: {df.shape}")

    # 🧹 Handle missing values
    missing_count = df.isnull().sum().sum()
    if missing_count > 0:
        df.dropna(inplace=True)
        print(f"🧹 Dropped rows with missing values: {missing_count}")
    else:
        print("✅ No missing values found")

    # 🧹 Remove duplicate rows
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        df.drop_duplicates(inplace=True)
        print(f"🧹 Dropped duplicate rows: {duplicate_count}")
    else:
        print("✅ No duplicates found")

    # 🔠 Encode categorical columns (if any)
    categorical_cols = df.select_dtypes(include='object').columns.tolist()
    if categorical_cols:
        df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
        print(f"🔠 Encoded categorical columns: {categorical_cols}")
    else:
        print("✅ No categorical columns to encode")

    # 🎯 Separate features and target
    if 'HeartDisease' not in df.columns:
        raise ValueError("❌ 'HeartDisease' column not found in dataset")

    X = df.drop('HeartDisease', axis=1)
    y = df['HeartDisease']

    # 📊 Identify numeric columns
    numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
    non_numeric_cols = X.columns.difference(numeric_cols)

    # 📏 Apply scaling only on numeric columns
    scaler = StandardScaler()
    X_scaled_numeric = scaler.fit_transform(X[numeric_cols])

    # Convert scaled numeric features back to DataFrame
    X_scaled_df = pd.DataFrame(X_scaled_numeric, columns=numeric_cols, index=X.index)

    # 🧱 Combine scaled numeric and unscaled non-numeric features
    X_processed = pd.concat([X_scaled_df, X[non_numeric_cols]], axis=1)

    # 💾 Save the scaler
    os.makedirs('models', exist_ok=True)
    with open('models/scaler.pkl', 'wb') as f:
        pickle.dump(scaler, f)

    # 🌲 Feature Selection using Random Forest
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_processed, y)

    selector = SelectFromModel(rf, threshold='median', prefit=True)
    X_selected = selector.transform(X_processed)

    original_features = X_processed.columns.tolist()
    selected_features = np.array(original_features)[selector.get_support()].tolist()
    print(f"🎯 Selected {len(selected_features)} important features out of {len(original_features)}")

    # 💾 Save selected features
    with open('models/selected_features.txt', 'w') as f:
        for feat in selected_features:
            f.write(f"{feat}\n")

    # ✂️ Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        X_selected, y, test_size=0.2, stratify=y, random_state=42
    )

    return X_train, X_test, y_train, y_test, selected_features, scaler
````

####**code steps**


1. **Loads the data**  
   → Reads `heart.csv` from the `data` folder.

2. **Handles missing values**  
   → Removes rows that have any empty cells.

3. **Removes duplicate rows**  
   → Keeps only unique entries.

4. **Encodes text columns**  
   → Converts text data into numbers using one-hot encoding.

5. **Splits the data**  
   → Separates input features (`X`) and target column (`HeartDisease` → `y`).

6. **Scales the features**  
   → Makes sure all numerical columns are on the same scale using `StandardScaler`.

7. **Saves the scaler**  
   → Saves the scaler object to use later during prediction.

8. **Selects important features**  
   → Uses `RandomForestClassifier` to keep only the best features.

9. **Saves selected features**  
   → Stores the final feature names in a file for reference.

10. **Splits into train and test sets**  
    → Divides the data into 80% training and 20% testing for model building.

11. **Returns processed data**  
    → Gives back the train/test sets, selected features, and scaler object.

### **Train Model using ML algorithms**



This step is focused on training multiple machine learning models with **hyperparameter tuning**, **data balancing**, and **saving models** for reuse or deployment.


**1. Prepares the Training Environment**

- ✅ Creates a directory (`models/`) to store the trained models.
- ⚖️ Applies **SMOTE (Synthetic Minority Over-sampling Technique)** to balance the dataset, ensuring **equal representation** of both classes in `y_train`.

 **2. Defines Multiple Models to Train**

Three classifiers are defined:

- `XGBoostClassifier`
- `RandomForestClassifier`
- `DecisionTreeClassifier`

Each model is paired with its own **hyperparameter grid** for tuning using **GridSearchCV**, which applies **cross-validation** to find the best combination of parameters.

**3. Trains Models Using Grid Search**

For each model:

- Runs **`GridSearchCV`** to identify the best hyperparameters.
- Trains the model on the **balanced dataset** (`X_train_bal`, `y_train_bal`).
- Selects the **best estimator** based on **accuracy**.

**4. Saves the Trained Models Locally**

Each trained model is saved as a `.pkl` file in the `models/` folder for reuse.

- Example file paths:
  - `models/xgboost_model.pkl`
  - `models/randomforest_model.pkl`
  - `models/decisiontree_model.pkl`

**5. Final result**

- ✅ Trained and tuned versions of 3 different models.
- 💾 Models are **saved locally** in `.pkl` format for prediction or deployment later.

````python
# train_models.py
import os
import pickle
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

def train_models(X_train, y_train, model_dir="models"):
    os.makedirs(model_dir, exist_ok=True)

    smote = SMOTE(random_state=42)
    X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

    models = {
        "XGBoost": {
            "model": XGBClassifier(random_state=42, eval_metric='logloss'),
            "params": {
                'n_estimators': [100],
                'max_depth': [3, 5],
                'learning_rate': [0.05],
                'subsample': [0.8]
            }
        },
        "RandomForest": {
            "model": RandomForestClassifier(random_state=42),
            "params": {
                'n_estimators': [100],
                'max_depth': [5],
                'min_samples_split': [5]
            }
        },
        "DecisionTree": {
            "model": DecisionTreeClassifier(random_state=42),
            "params": {
                'max_depth': [3, 5],
                'min_samples_split': [2, 4]
            }
        }
    }

    for name, cfg in models.items():
        print(f"\n🔍 Training {name}...")
        grid = GridSearchCV(cfg["model"], cfg["params"], cv=3, scoring='accuracy', n_jobs=-1)
        grid.fit(X_train_bal, y_train_bal)
        best_model = grid.best_estimator_

        model_path = os.path.join(model_dir, f"{name.lower()}_model.pkl")
        with open(model_path, "wb") as f:
            pickle.dump(best_model, f)

        print(f"✅ {name} saved to {model_path} with best params: {grid.best_params_}")
````

````python
# evaluate_models.py
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    classification_report, roc_curve, auc
)
import joblib

graph_dir = "model_graphs"
os.makedirs(graph_dir, exist_ok=True)

def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    path = os.path.join(graph_dir, f"{model_name}_confusion_matrix.png")
    plt.savefig(path)
    plt.close()
    return path

def plot_roc_curve(y_true, y_prob, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f'ROC Curve - {model_name}')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    path = os.path.join(graph_dir, f"{model_name}_roc_curve.png")
    plt.savefig(path)
    plt.close()
    return path

def evaluate_models(X_test, y_test, model_path='models'):
    model_files = [f for f in os.listdir(model_path) if f.endswith('_model.pkl')]
    results = {}
    models_dict = {}

    for model_file in model_files:
        model_name = model_file.replace('_model.pkl', '').capitalize()
        print(f"\n📦 Loading {model_name}...")
        with open(os.path.join(model_path, model_file), 'rb') as f:
            model = pickle.load(f)
            models_dict[model_name] = model

        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred)
        rec = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc_score = roc_auc_score(y_test, y_prob)

        print(f"📊 Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, AUC: {auc_score:.4f}")
        print("📋 Classification Report:")
        print(classification_report(y_test, y_pred))

        plot_confusion_matrix(y_test, y_pred, model_name)
        plot_roc_curve(y_test, y_prob, model_name)

        results[model_name] = {
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1_score": f1,
            "roc_auc": auc_score
        }

    print("\n✅ All models evaluated!")

    # 🔍 Select Best Model Based on Accuracy (can change to 'roc_auc' or 'f1_score')
    best_model_name = max(results, key=lambda x: results[x]['accuracy'])
    best_model = models_dict[best_model_name]
    best_model_path = os.path.join(model_path, "best_model.pkl")
    joblib.dump(best_model, best_model_path)

    print(f"\n🏆 Best Model: {best_model_name}")
    print(f"✅ Saved best model to: {best_model_path}")

    return results
````

#### 📌 **MLflow Logging Explained**



- `import mlflow`  
  → Loads MLflow to enable experiment tracking.

- `import mlflow.sklearn`  
  → Loads support for logging scikit-learn models.

- `mlflow.log_params(grid.best_params_)`  
  → Logs the best hyperparameters found during model training.

- `signature = infer_signature(X_train_bal, best_model.predict(X_train_bal))`  
  → Detects and records the model's expected input and output format.

- `mlflow.sklearn.log_model(...)`  
  → Saves the trained model to MLflow, along with:
  - Sample input
  - Input/output signature
  - Optionally registers it with a model name for version control


````python
# train_models.py
import os
import pickle
import mlflow
import mlflow.sklearn
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from mlflow.models.signature import infer_signature

def train_models(X_train, y_train, model_dir="models"):
    os.makedirs(model_dir, exist_ok=True)

    smote = SMOTE(random_state=42)
    X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

    models = {
        "XGBoost": {
            "model": XGBClassifier(random_state=42, eval_metric='logloss'),
            "params": {
                'n_estimators': [100],
                'max_depth': [3, 5],
                'learning_rate': [0.05],
                'subsample': [0.8]
            }
        },
        "RandomForest": {
            "model": RandomForestClassifier(random_state=42),
            "params": {
                'n_estimators': [100],
                'max_depth': [5],
                'min_samples_split': [5]
            }
        },
        "DecisionTree": {
            "model": DecisionTreeClassifier(random_state=42),
            "params": {
                'max_depth': [3, 5],
                'min_samples_split': [2, 4]
            }
        }
    }

    for name, cfg in models.items():
        print(f"\n🔍 Training {name}...")
        with mlflow.start_run(run_name=f"{name}_training"):
            grid = GridSearchCV(cfg["model"], cfg["params"], cv=3, scoring='accuracy', n_jobs=-1)
            grid.fit(X_train_bal, y_train_bal)
            best_model = grid.best_estimator_

            mlflow.log_params(grid.best_params_)

            input_example = X_train_bal[:1]
            signature = infer_signature(X_train_bal, best_model.predict(X_train_bal))

            mlflow.sklearn.log_model(
                sk_model=best_model,
                artifact_path=f"{name}_model",
                input_example=input_example,
                signature=signature,
                registered_model_name=f"HeartDisease_{name}"  # Register here
            )

            model_path = os.path.join(model_dir, f"{name.lower()}_model.pkl")
            with open(model_path, "wb") as f:
                pickle.dump(best_model, f)

            print(f"✅ {name} saved with best params: {grid.best_params_}")
````

````python
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import mlflow
import mlflow.sklearn
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    classification_report, roc_curve, auc
)
from mlflow.models.signature import infer_signature
import joblib

graph_dir = "model_graphs"
os.makedirs(graph_dir, exist_ok=True)

def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    path = os.path.join(graph_dir, f"{model_name}_confusion_matrix.png")
    plt.savefig(path)
    plt.close()
    return path

def plot_roc_curve(y_true, y_prob, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f'ROC Curve - {model_name}')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    path = os.path.join(graph_dir, f"{model_name}_roc_curve.png")
    plt.savefig(path)
    plt.close()
    return path

def evaluate_models(X_test, y_test, model_path='models'):
    model_files = [f for f in os.listdir(model_path) if f.endswith('_model.pkl')]
    results = {}
    models_dict = {}

    for model_file in model_files:
        model_name = model_file.replace('_model.pkl', '').capitalize()
        print(f"\n📦 Loading {model_name}...")
        with open(os.path.join(model_path, model_file), 'rb') as f:
            model = pickle.load(f)
            models_dict[model_name] = model

        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred)
        rec = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc_score = roc_auc_score(y_test, y_prob)

        print(f"📊 Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, AUC: {auc_score:.4f}")
        print("📋 Classification Report:")
        print(classification_report(y_test, y_pred))

        cm_path = plot_confusion_matrix(y_test, y_pred, model_name)
        roc_path = plot_roc_curve(y_test, y_prob, model_name)

        input_example = X_test[:1]
        signature = infer_signature(X_test, model.predict(X_test))

        with mlflow.start_run(run_name=f"{model_name}_evaluation"):
            mlflow.log_param("model_name", model_name)
            mlflow.log_metric("accuracy", acc)
            mlflow.log_metric("precision", prec)
            mlflow.log_metric("recall", rec)
            mlflow.log_metric("f1_score", f1)
            mlflow.log_metric("roc_auc", auc_score)
            mlflow.log_artifact(cm_path, artifact_path="plots")
            mlflow.log_artifact(roc_path, artifact_path="plots")
            mlflow.sklearn.log_model(
                sk_model=model,
                name=f"{model_name}_evaluated",
                input_example=input_example,
                signature=signature
            )

        results[model_name] = {
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1_score": f1,
            "roc_auc": auc_score
        }

    print("\n✅ All models evaluated and metrics logged to MLflow!")

    # 🔍 Select Best Model Based on Accuracy (or change to 'roc_auc' or 'f1_score')
    best_model_name = max(results, key=lambda x: results[x]['accuracy'])
    best_model = models_dict[best_model_name]
    best_model_path = os.path.join(model_path, "best_model.pkl")
    joblib.dump(best_model, best_model_path)

    print(f"\n🏆 Best Model: {best_model_name}")
    print(f"✅ Saved best model to: {best_model_path}")

    return results
````

**Registering trained models in MLflow’s Model Registry:**

- After training a model and logging it with MLflow, you can register it to the Model Registry.  
- Registration gives the model a unique name and tracks its versions.  
- This creates a centralized catalog of your models, enabling easy version control.  

**Managing multiple versions of models:**

- Every time you log a new trained model and register it under the same name, MLflow creates a new version.  
- You can compare different versions to see which performs best.  
- Versions help keep track of model evolution over time.  

**Promoting the best model version to the Production stage:**

- After evaluating different model versions, you can promote the best one to the Production stage.  
- Production stage indicates this model is ready for deployment and real-world use.  
- You can also have other stages like Staging or Archived for different lifecycle phases.  

**Tracking experiments, versions, and metadata using MLflow UI:**

- MLflow UI provides an interface to visualize experiments, model versions, stages, metrics, parameters, and artifacts.  
- It allows you to track the full lifecycle of your ML models from training to deployment.  
- You can review run histories, compare metrics, and manage model stage transitions visually.  



### **🔁 Model Retraining – What It Is & Why It Matters**

#### **What is Model Retraining?**



**What is Model Retraining?**

**Model retraining** is the process of updating a machine learning model by training it again on **new data** (or a mix of old and new data) to maintain or improve its predictive performance over time.



#### **Why is Model Retraining Important?**

| Reason                        | Explanation                                                                 |
|-------------------------------|-----------------------------------------------------------------------------|
| **1. Data Changes Over Time** | Real-world behavior shifts, making old data outdated.                      |
| **2. Concept Drift**          | The relationship between input features and target variables changes.      |
| **3. Maintain Accuracy**      | Model performance naturally decays over time without updates.              |
| **4. New Data Available**     | Retraining allows the model to learn from new patterns.                    |
| **5. Business Adaptability**  | Models need to reflect current user needs, trends, or market conditions.   |
| **6. Compliance Requirements**| Some industries require periodic retraining for fairness and accuracy.     |

#### **When Should You Retrain a Model?**

- When **new data** is collected regularly.
- When the **model performance drops** (e.g., lower accuracy or AUC).
- When **data drift** or **concept drift** is detected.
- On a **scheduled basis** (e.g., weekly/monthly).
- After **feature engineering** or **data schema changes**.






#### **Real-World Examples**



| Domain        | Why Retraining is Needed                                  |
|---------------|-----------------------------------------------------------|
| E-commerce     | Customer behavior changes seasonally.                    |
| Healthcare     | New disease patterns emerge (e.g., COVID variants).      |
| Finance        | Fraud tactics evolve to bypass detection.               |
| Social Media   | Trends and language change rapidly.                      |

### **Best Practices and Tips**


1. **Clearly Define the Problem**  
   Start with a well-scoped **problem statement** that aligns with business goals. Ambiguity in objectives can lead to ineffective models.

2. **Understand Your Data**  
   Spend time on **EDA (Exploratory Data Analysis)** to uncover patterns, anomalies, and relationships. Good data understanding leads to better features.

3. **Maintain Data Quality**  
   Always ensure your data is **clean**, **balanced**, and **relevant**. Garbage in = garbage out.

4. **Track Everything**  
   Use tools like **MLflow** to track **experiments**, **metrics**, **parameters**, and **model versions**. This improves reproducibility and transparency.

5. **Validate Properly**  
   Use techniques like **cross-validation** to ensure your model generalizes well to unseen data.

6. **Retrain Regularly**  
   Incorporate **model retraining pipelines** to handle **data drift** and keep models up-to-date.

7. **Modularize Your Code**  
   Break your pipeline into reusable components (e.g., preprocessing, training, evaluation). This increases scalability and maintainability.

8. **Document Your Work**  
   Maintain clear documentation of your code, data sources, assumptions, and results for better collaboration and future updates.

9. **Avoid Overfitting**  
   Regularly monitor training vs. validation performance and apply techniques like **regularization**, **dropout**, or **early stopping**.

10. **Focus on Interpretability**  
   Use tools like **SHAP**, **LIME**, or feature importance plots to interpret your model and build trust with stakeholders.



## **Shortcomings**



1. **Data Dependency**  
   ML models are only as good as the data they are trained on. Poor-quality or biased data can lead to inaccurate predictions.

2. **Model Degradation Over Time**  
   Due to **data drift** or changing patterns, models may lose accuracy if not **retrained regularly**.

3. **Overfitting and Underfitting**  
   Balancing model complexity is challenging. Overfitting results in poor generalization, while underfitting leads to low accuracy.

4. **Lack of Interpretability**  
   Complex models like deep neural networks may lack transparency, making it hard to explain decisions to stakeholders.

5. **Infrastructure Limitations**  
   Building and maintaining end-to-end ML systems often require scalable infrastructure, which can be costly and complex to manage.

6. **Experiment Tracking Gaps**  
   Without proper tools (e.g., **MLflow**), managing multiple experiments and model versions can become chaotic and error-prone.

7. **Reproducibility Challenges**  
   Failing to track **data versions**, **random seeds**, or **dependencies** can make it difficult to reproduce results.

8. **Ethical and Bias Concerns**  
   Unchecked ML models can reinforce existing biases, leading to unfair or harmful outcomes.

9. **Deployment and Monitoring Complexity**  
   Taking a model from notebook to real-world application involves many moving parts—APIs, CI/CD, monitoring—that can introduce errors.

10. **High Maintenance Costs**  
   Continuous retraining, monitoring, and data management require time, effort, and resources, which can be burdensome at scale.
