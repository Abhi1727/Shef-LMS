# 🤖 **What is Automation?**

#**Automated COVID-19 Global Metrics Pipeline**

## **Problem Statement**
Monitoring the ever-changing COVID-19 landscape requires frequent data retrieval, processing, and visualization. Manually performing these tasks is time-consuming and error-prone. By automating the workflow—from data acquisition through KPI computation to chart generation—we ensure up-to-date, reproducible insights with minimal human intervention.

## **Objectives**
1. **Automate Data Fetching**
Retrieve the latest COVID-19 statistics for all countries via a public API.

2. **Compute Global KPIs**
Calculate key performance indicators such as total cases, deaths, recoveries, active cases, mortality rate, and recovery rate.

3. **Persist Results**
Save the KPI summary table to a CSV for archival or downstream analysis.

4. **Identify Top-10 Impacted Countries**
Determine which countries have the highest case counts and compute their individual death rates.

5. **Generate Visualizations**
Produce bar charts for total cases, total recoveries, and death rates of the top-10 countries, and save them as a PNG.

6. **Enable Hands-Off Operation with Cron Scheduling**
To ensure the process runs automatically without manual intervention, wrap all necessary operations into a single script ( Python script).
This script should handle all tasks, including environment setup, execution logic, and logging. Once the script is prepared and made executable, schedule it using a system scheduler like **cron**.
This can be done by adding an entry to the system's crontab (using crontab -e) specifying when and how frequently the script should run. For example, to run the script daily at 2:00 AM:

**0 2 * * * /absolute/path/to/your_script.sh >> /path/to/logfile.log 2>&1**

This setup ensures the process is executed automatically at the specified time, making the workflow fully automated and reliable without requiring manual triggers.



## **Step 1: Import libraries**
Load the Python modules needed for filesystem operations, HTTP requests, data handling, and plotting.

````python
import os
import requests
import pandas as pd
import matplotlib.pyplot as plt
````

## **Step 2: Prepare the output directory**
Ensure there’s a plots/ folder to save our CSV and PNG files.

````python
os.makedirs("plots", exist_ok=True)
````

## **Step 3: Fetch data from the API**
Request the latest per-country COVID-19 stats and parse the JSON response.

````python
url = "https://disease.sh/v3/covid-19/countries"
resp = requests.get(url)
resp.raise_for_status()
data = resp.json()
print("✅ Fetched COVID-19 data")
````

## **Step 4: Build a pandas DataFrame**
Convert the JSON into a DataFrame containing only the key columns we need.

````python
df = pd.DataFrame(data)[['country', 'cases', 'deaths', 'recovered', 'active']]
````

# **Step 5: Compute global KPIs**
Sum across all countries for totals, then derive mortality and recovery rates.

````python
total_cases     = df['cases'].sum()
total_deaths    = df['deaths'].sum()
total_recovered = df['recovered'].sum()
total_active    = df['active'].sum()

mortality_rate = round((total_deaths / total_cases) * 100, 2)
recovery_rate  = round((total_recovered / total_cases) * 100, 2)

kpi = {
    "Total Cases":        total_cases,
    "Total Deaths":       total_deaths,
    "Total Recovered":    total_recovered,
    "Total Active":       total_active,
    "Mortality Rate (%)": f"{mortality_rate}%",
    "Recovery Rate (%)":  f"{recovery_rate}%"
}
````

# **Step 6: Save KPI table to CSV**
Write the KPI dictionary into a two-column CSV (Metric,Value) under plots/.

````python
df_kpis = pd.DataFrame.from_dict(kpi, orient='index', columns=['Value'])
df_kpis.index.name = 'Metric'
df_kpis.to_csv("plots/global_covid_kpis.csv")
print("✅ Saved KPI table to plots/global_covid_kpis.csv")
````

##**Step 7: Identify top 10 countries by case count**
Compute each country’s death rate, sort by total cases, and select the top 10.

````python
top10 = (
    df
    .assign(death_rate=(df['deaths'] / df['cases'] * 100).round(2))
    .sort_values('cases', ascending=False)
    .head(10)
    .set_index('country')
    [['cases', 'recovered', 'death_rate']]
)
top10.columns = ['Total Cases', 'Total Recovered', 'Death Rate (%)']
````

# **Step 8: Plot bar charts for the top 10 metrics**
Create side-by-side bar charts for total cases, total recovered, and death rate, then save as a PNG.

````python
fig, axes = plt.subplots(1, 3, figsize=(18,5), sharex=True)

top10['Total Cases'].plot.bar(ax=axes[0])
axes[0].set_title("Total Cases")
axes[0].tick_params(axis='x', rotation=45)

top10['Total Recovered'].plot.bar(ax=axes[1])
axes[1].set_title("Total Recovered")
axes[1].tick_params(axis='x', rotation=45)

top10['Death Rate (%)'].plot.bar(ax=axes[2])
axes[2].set_title("Death Rate (%)")
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig("plots/top10_covid_metrics.png", dpi=100)
plt.close()
print("✅ Saved top-10 country charts to plots/top10_covid_metrics.png")
````

````python
import os                          # access operating system functionality (e.g., file and directory operations)
import requests                    # send HTTP requests to external APIs
import pandas as pd                # data manipulation and analysis library
import matplotlib.pyplot as plt    # plotting library for creating visualizations

# 1) Prepare output directory
os.makedirs("plots", exist_ok=True)  # create a folder named "plots" if it doesn’t already exist

try:
    # 2) Fetch data
    url = "https://disease.sh/v3/covid-19/countries"  # API endpoint for per‐country COVID-19 data
    resp = requests.get(url)                          # send GET request to the API
    resp.raise_for_status()                           # raise an error if the HTTP request failed
    data = resp.json()                                # parse the JSON response into Python objects
    print("✅ Fetched COVID-19 data")                 # log success

    # 3) Build DataFrame
    df = pd.DataFrame(data)[                          # convert the JSON list to a DataFrame
        ['country','cases','deaths','recovered','active']  # select only the needed columns
    ]

    # 4) Compute global KPIs
    total_cases     = df['cases'].sum()               # sum of all country case counts
    total_deaths    = df['deaths'].sum()              # sum of all country death counts
    total_recovered = df['recovered'].sum()           # sum of all country recovered counts
    total_active    = df['active'].sum()              # sum of all country active counts
    mortality_rate  = round((total_deaths / total_cases) * 100, 2)  # global mortality percentage
    recovery_rate   = round((total_recovered / total_cases) * 100, 2)  # global recovery percentage

    kpi = {                                           # assemble a dictionary of KPIs
        "Total Cases":        total_cases,
        "Total Deaths":       total_deaths,
        "Total Recovered":    total_recovered,
        "Total Active":       total_active,
        "Mortality Rate (%)": f"{mortality_rate}%",
        "Recovery Rate (%)":  f"{recovery_rate}%"
    }

    # 5) Save KPI table to CSV in plots/
    df_kpis = pd.DataFrame.from_dict(                 # create DataFrame from KPI dict
        kpi, orient='index', columns=['Value']
    )
    df_kpis.index.name = 'Metric'                     # name the index column "Metric"
    kpi_csv = "plots/global_covid_kpis.csv"           # define output CSV path
    df_kpis.to_csv(kpi_csv)                           # write KPIs to CSV
    print(f"✅ Saved KPI table to {kpi_csv}")          # log success

    # 6) Top 10 countries by cases
    top10 = (
        df
        .assign(death_rate=(df['deaths']/df['cases']*100).round(2))  # calculate each country’s death rate
        .sort_values('cases', ascending=False)                      # sort countries descending by case count
        .head(10)                                                   # select the top 10
        .set_index('country')                                       # set country names as the index
        [['cases','recovered','death_rate']]                        # choose columns to keep
    )
    top10.columns = ['Total Cases','Total Recovered','Death Rate (%)']  # rename columns for clarity

    # 7) Plot bar charts and save
    fig, axes = plt.subplots(1, 3, figsize=(18,5), sharex=True)  # create a 1×3 grid of plots

    top10['Total Cases'].plot.bar(ax=axes[0])       # bar chart of total cases
    axes[0].set_title("Total Cases")                # title for first subplot
    axes[0].tick_params(axis='x', rotation=45)      # rotate x-axis labels for readability

    top10['Total Recovered'].plot.bar(ax=axes[1])   # bar chart of total recovered
    axes[1].set_title("Total Recovered")            # title for second subplot
    axes[1].tick_params(axis='x', rotation=45)      # rotate x-axis labels

    top10['Death Rate (%)'].plot.bar(ax=axes[2])    # bar chart of death rate
    axes[2].set_title("Death Rate (%)")             # title for third subplot
    axes[2].tick_params(axis='x', rotation=45)      # rotate x-axis labels

    plt.tight_layout()                              # adjust spacing to prevent overlap
    plot_png = "plots/top10_covid_metrics.png"      # define output PNG path
    plt.savefig(plot_png, dpi=100)                  # save the figure to file
    plt.close()                                     # close the figure to free memory
    print(f"✅ Saved top-10 country charts to {plot_png}")  # log success

except Exception as e:
    print(f"❌ Error: {e}")                         # catch and log any errors
````

## **What is Cron**

**Cron** is a time-based job scheduler daemon on Unix-like systems. It runs in the background and checks every minute for tasks to execute.


### Crontab

The **crontab** (cron table) is a per-user text file listing scheduled jobs. Each line consists of:


- **minute** (0–59)  
- **hour** (0–23)  
- **day-of-month** (1–31)  
- **month** (1–12 or Jan–Dec)  
- **day-of-week** (0–7 or Sun–Sat, where both 0 and 7 mean Sunday)  
- **command** the script or program to run

Common crontab commands:

- `crontab -e` — edit your crontab  
- `crontab -l` — list your scheduled jobs  
- `crontab -r` — remove your crontab


### Why use cron/crontab?

- **Automation**  
  Execute scripts or commands at fixed intervals without manual intervention.  
- **Reliability**  
  Cron jobs run whether you’re logged in or not, and persist across reboots.  
- **Flexibility**  
  Schedule anything from every minute to once a year, and capture output or errors to log files.  
- **Low overhead**  
  Built into nearly every Linux and macOS environment—no extra services required.

# Schedule Your Python Script in VS Code with Cron (Every 2 Minutes)

1. **Open Your Project in VS Code**  
   - Launch VS Code and open the folder containing your script (e.g. `covid_report.py`).

2. **Verify Your Script Runs with Python 3**  
   - In VS Code’s integrated terminal, run:  
     ```bash
     python3 /full/path/to/your/project/covid_report.py
     ```  
   - Confirm it executes without error and populates `plots/` (e.g. `global_covid_kpis.csv`, `top10_covid_metrics.png`).

3. **Open Your Crontab for Editing**  
   - In the same terminal, type:  
     ```bash
     crontab -e
     ```  
   - If prompted, choose **nano** or **vim** as your editor.

4. **Add the “Every 2 Minutes” Entry**  
   - Append this line at the bottom (replace paths accordingly):  
     ```cron
     */2 * * * * python3 /full/path/to/your/project/app.py >> /full/path/to/your/project/plots/app.log 2>&1
     ```

1. **`*/2 * * * *`**  
   - **Minute field (`*/2`)**: run every 2 minutes (i.e. at 0, 2, 4, …, 58).  
   - **Hour (`*`)**: every hour.  
   - **Day of month (`*`)**: every day.  
   - **Month (`*`)**: every month.  
   - **Day of week (`*`)**: every weekday.  

2. **`python3 /full/path/to/your/project/app.py`**  
   - Invokes your script with the system’s Python 3 interpreter.  
   - Replace `/full/path/to/your/project/app.py` with the absolute path to your script file.

3. **`>> /full/path/to/your/project/plots/app.log`**  
   - Appends standard output (stdout) to `app.log` in your `plots/` folder.  
   - If `app.log` doesn’t exist, cron creates it automatically.

4. **`2>&1`**  
   - Redirects standard error (stderr, descriptor 2) to the same place as stdout (descriptor 1).  
   - Ensures that both normal messages and error messages get logged into `app.log`.

## ✅ What is a Webhook?

A **Webhook** is a way for one application to send **real-time data** to another application as soon as an event occurs.

It works like a **reverse API**.

> 🔄 Instead of requesting data (like APIs do), webhooks **automatically send data** when a specific event happens.

### 📌 Example:
Imagine you’re using a payment gateway like **Razorpay** or **Stripe**.  
When a **payment is successful**, a **webhook** sends that **payment info to your server instantly**.

## 🍕 Real Example: Food Order and Webhook

### 📦 Scenario: You're running a cloud kitchen that partners with **Zomato**

---

### 🧩 Without Webhook (Using API Only)
You’d have to:

- Keep polling Zomato’s API every few seconds:  
  *“Any new orders?”*  
  *“Any new orders?”*

❗ This adds load on your server and can cause **delays**.

---

### ⚡ With Webhook
Instead, Zomato offers a **webhook**.

You set up your server to accept webhook notifications:  
`https://mycloudkitchen.com/zomato-webhook`

Now, **whenever a customer places an order on Zomato**:

➡️ Zomato **automatically sends the order details** to your webhook URL **instantly**.

✅ You don’t have to check — it **pushes** the data to you.

---

### 🔁 What Happens Behind the Scenes

1. 🍽️ A customer orders a pizza on Zomato.  
2. 🧠 Zomato’s system detects the event: **"New Order Created"**  
3. 📡 Zomato sends a **POST request** to your webhook URL with all the order details (items, address, etc.).  
4. 🧾 Your server receives the data and:
   - Notifies the chef 👨‍🍳  
   - Prints a receipt 🖨️  
   - Starts the cooking timer automatically ⏱️  

---

### ✅ Benefits

- ⚡ **Instant order processing**
- 🚫 **Zero delay or polling**
- 🔄 **Real-time kitchen automation**
- 😊 **Better customer experience**


## ❓ Why is a Webhook Required?

**Webhooks** are useful when:

- ⏱️ You need **real-time updates** without polling the API.
- ⚙️ You want to **automate workflows** when something happens  
  (e.g., order placed, user signup).
- 📉 You need to **reduce server load**  
  (no need to keep asking the API *"has anything changed?"*).

---

### 🛠️ Common Use Cases

## 🔄 API vs Webhook – Key Comparison Table

| **Feature**        | **API (Application Programming Interface)**               | **Webhook**                                            |
|--------------------|-----------------------------------------------------------|--------------------------------------------------------|
| **Data Flow**       | Client pulls data from the server                         | Server pushes data to the client                       |
| **Trigger**         | Client must make a request (polling or on-demand)         | Event on server automatically triggers a request       |
| **Speed**           | Can be slower due to polling                              | Instant and real-time                                  |
| **Use Case**        | When you want to get or manipulate data actively          | When you want to receive updates automatically         |
| **Control**         | Client controls when data is fetched                      | Server decides when to send data                       |
| **Example Use**     | Get user details, fetch transactions, create new posts    | Payment success alerts, GitHub push notifications      |
| **Complexity**      | More flexible but can require frequent polling            | Simple and lightweight for specific triggers           |
| **Resource Usage**  | More load on both client and server                       | More efficient, no need for constant polling           |

|--------------------|--------------------------------------------------|-----------------------------------------------------|
| **Food delivery**   | You keep calling the restaurant to ask if the food is ready | The restaurant calls you when the food is ready     |
| **Email notification** | You check your inbox every 5 mins              | You get a notification the moment you receive an email |

## 💡 When to Use What?

### ✅ Use **API** when:
- You need **full control** over the data
- You want to **fetch data anytime**
- You need **CRUD operations** (Create, Read, Update, Delete)

### ⚡ Use **Webhook** when:
- You need **instant notifications**
- You want to **automate workflows**
- You want to **reduce polling and server load**

## 🧠 Webhooks and Data Science – What's the Role?

### 🚫 Webhook Creation ≠ Data Science Task

Creating or configuring a webhook (like defining the endpoint, handling authentication, or setting up the server route) is generally part of **software development** or **backend engineering**.

- It involves:
  - Writing server-side code to handle incoming requests
  - Managing infrastructure and routing
  - Ensuring security and validation

### ✅ Triggering a Webhook = Data Science Relevant

While creation is not a DS responsibility, **triggering a webhook** (i.e., sending a `POST` request to a webhook URL) **can be part of a Data Scientist's workflow**, especially when automating:

- Model deployment notifications  
- Batch job completions  
- Triggering downstream processes (e.g., alerting a service after predictions are ready)

### 📌 Analogy:
Using a webhook is like using an API —  
📤 **You send a POST request with data**  
🧠 As a DS, if your script/model needs to notify another system, you **can trigger the webhook**.

> After a machine learning model completes prediction on a large dataset, the DS script sends a `POST` request to a webhook URL to trigger a reporting or visualization pipeline.

# **✅ Summary ✅**

## 📚 What Did We Learn?

- **Workflow Management**: Designing, running, and monitoring multi-step processes to keep tasks organized and efficient.  
- **Automation**: Using scripts or tools to handle repetitive work without manual effort.  
- **Benefits of Automation**: Saves time, reduces errors, and ensures consistency.  
- **Automation in Data Science & Analysis**: Automates data fetching, cleaning, modeling, reporting, and monitoring.  
- **Example Use Cases**:  
  - Daily COVID-19 or stock data collection  
  - Hands-free data cleaning  
  - Auto-updating dashboards  
  - Scheduled report delivery  
  - Real-time alerts on model performance  
- **Common Automation Tools**: Python scripts, cron jobs

## ⚠️ Shortcomings

- **Initial Setup Overhead**  
  Requires time and effort to design workflows and configure tools before benefits are realized.

- **Maintenance Burden**  
  Automated pipelines need regular updates and monitoring to handle changing data sources or dependencies.

- **Debugging Complexity**  
  Failures in automated tasks can be harder to diagnose without proper logging and alerting.

- **Over-Automation Risk**  
  Automating every task can lead to rigid processes that are difficult to adapt to one-off or ad-hoc needs.

- **Security Concerns**  
  Storing credentials or exposing endpoints (e.g., webhooks) demands careful handling to prevent unauthorized access.


## ✅ Best Practices

- **Modularize Your Workflows**  
  Break pipelines into discrete, reusable steps (e.g., fetch → transform → load) to simplify maintenance and testing.

- **Implement Idempotency**  
  Design tasks so they can safely run multiple times without producing duplicate side effects (e.g., use upserts or timestamp checks).

- **Use Robust Logging & Monitoring**  
  Capture both stdout and stderr in log files, and integrate alerts (email/Slack) to catch failures or performance degradation early.

- **Version Control Everything**  
  Store scripts, configs, and crontab entries in Git (or similar) to track changes, enable rollbacks, and support collaboration.

- **Secure Your Credentials**  
  Never hard-code API keys or passwords; use environment variables, encrypted vaults, or secret managers.

- **Parameterize & Document**  
  Keep schedules, file paths, and thresholds configurable; accompany each pipeline with clear README instructions and inline comments.

- **Test & Validate**  
  Develop unit and integration tests for critical steps (data validation, API responses, plotting functions) before deploying.

- **Graceful Error Handling**  
  Catch exceptions in scripts, write meaningful error messages, and ensure downstream tasks aren’t left in a broken state.
