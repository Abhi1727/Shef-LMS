# **Introduction to Experimental Design**

## **What is Experimental Design?** 🧪📊

Experimental design is like planning a science experiment 🧑‍🔬 where you decide how to test a question by controlling factors so you can find out cause and effect relationships. It’s your experiment’s roadmap!

## **Importance in Data Science and Research** 💡📈

Say you want to see if a new app feature 📱 increases user engagement. Experimental design helps you set up the test so you can confidently say yes or no about its impact — not just guess.

**Good design helps you:**

- Minimize bias 🎯  
- Control variability 🔄  
- Draw meaningful conclusions 🧠  
- Make results reproducible 🔁  

## **Types of Variables**

##### **Independent Variables (IV)** 🎛️  
The factor you change on purpose.  
*Example:* Testing different fertilizers 🌱. Fertilizer type = IV.

##### **Dependent Variables (DV)**📏  
The outcome you measure, which depends on the IV.  
*Example:* Plant height after applying fertilizers.

##### **Control Variables** 🔒  
Factors you keep constant so they don’t confuse results.  
*Example:* Same sunlight ☀️, water 💧, soil type.

## **Types of Experimental Designs**

#### **Completely Randomized Design (CRD)** 🎲  
Randomly assign units to treatments with no restrictions.  
*Example:* Students are randomly given either a new teaching method 🧑‍🏫 or the traditional method.

#### **Randomized Block Design (RBD)** 🧱  
Group subjects into blocks by a variable, then randomly assign treatments within each block.  
*Example:* Testing a drug 💊, blocking by age groups 👵🧑👦, then randomizing treatments within each age block.

#### **Factorial Design** 🔀  
Test multiple factors at once with all possible combinations.  
*Example:* Test fertilizer type (A, B) 🌿 and watering frequency (daily, weekly) 💧. Try all 4 combos.

## **Principles of Experimental Design**

#### **Randomization** 🎰  
Randomly assign treatments to reduce bias.  
*Example:* Flipping a coin to assign treatments.

#### **Replication** 🔁  
Repeat the experiment on multiple subjects for reliability.  
*Example:* Testing fertilizer on 30 plants, not just one.

#### **Control** 🛑  
Keep other factors constant or use controls to isolate effects.  
*Example:* A group of plants gets no fertilizer to compare.

#### **Blocking** 🧱  
Group similar units to reduce variability.  
*Example:* Group patients by gender before assigning treatments.

## **Real Example**

**Scenario:** A coffee shop ☕ wants to test if playing different types of music 🎶 affects sales 💵.

- **IV:** Music type (Classical 🎻, Pop 🎤, No music 🚫)  
- **DV:** Sales revenue 💰  
- **Control Variables:** Time of day ⏰, day of week 📅, same staff 👩‍🍳  

##### **Designs**:

- **CRD:** Randomly pick days to play each music type.  
- **RBD:** Block by day of the week because sales vary on weekends vs weekdays, then randomly assign music within those blocks.  
- **Factorial:** Add volume levels (low 🔈, high 🔊) and test all combos.

### **Principles used**:

- **Randomization 🎲:** Random music assignments to days.  
- **Replication 🔂:** Test each music type multiple days.  
- **Control 🔒:** Keep staff and time fixed.  
- **Blocking 🧱:** Group days by weekday/weekend.

# **Hypothesis Testing Basics** 🧪📊

## **Understanding Hypotheses** 🤔

When you test something scientifically, you start with two statements called hypotheses:

##### **Null Hypothesis (H₀) ❌**  
This is the "no effect" or "no difference" statement. It assumes nothing new is happening.  
*Example:* A new drug has no effect on blood pressure.

##### **Alternative Hypothesis (H₁) ✔️**  
This is what you want to prove — that there is an effect or difference.  
*Example:* The new drug does lower blood pressure.

## **Types of Errors** ⚠️

When making decisions, you can make mistakes:

##### **Type I Error (False Positive) 🚨**  
You reject the null hypothesis (say there's an effect) when in reality, there isn’t one.  
Like a false alarm.  
*Example:* Saying the drug works when it actually doesn’t.

##### **Type II Error (False Negative) 🤦**  
You fail to reject the null hypothesis (say no effect) when there actually is one.  
Missing a real effect.  
*Example:* Saying the drug doesn’t work when it actually does.

## **Statistical Significance** 🎯

This tells you how strong your evidence is against the null hypothesis.

##### **p-value 📉**  
The probability of observing your data, or something more extreme, if the null hypothesis is true.

- Small p-value (usually < 0.05) → strong evidence against H₀ → reject H₀  
- Large p-value → weak evidence → fail to reject H₀  

##### **Alpha Level (α) 🔥**  
The threshold you set for significance (common choice: 0.05).  
It’s the maximum risk you accept of making a Type I error.  
If p-value < α, reject the null.



# **Introduction to Parametric Tests**

Parametric tests are statistical tests that assume your data comes from a population that follows a specific distribution, usually a normal distribution (bell curve). These tests use parameters like mean and standard deviation to make inferences.

## **Assumptions of Parametric Tests** ✅

### **Normality** 🧠  
Data should be approximately normally distributed.  
*Example:* Heights of people often follow a normal distribution.

### **Homoscedasticity (Equal Variance)** ⚖️  
Groups being compared should have similar variances.

### **Scale** 📏  
Data should be measured at least on an interval scale (meaningful differences between values).


# **One-Sample t-test** 🧪

## **What is it?**  
A test to compare the mean of one sample against a known or hypothesized population mean.

## **Why use it?**  
To check if your sample’s average is significantly different from a specific value.

## **Real Case (Framed Question)**  
A coffee shop claims the average temperature of their brewed coffee is 65°C. You measure 30 cups and want to know:  
*"Is the average coffee temperature actually different from 65°C?"*


```python
import numpy as np
from scipy import stats

# Sample coffee temperatures in °C
sample_temps = np.array([64, 66, 67, 65, 64, 63, 66, 65, 67, 66, 64, 65, 63, 65, 67, 66, 64, 66, 65, 63, 64, 65, 67, 66, 64, 65, 63, 65, 67, 66])

# Claimed average temperature
popmean = 65

# Run one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample_temps, popmean)

print(f"t-statistic: {t_stat:.3f}, p-value: {p_value:.3f}")
```

## **Insights**

- If p-value < 0.05, **reject the claim** → The coffee temperature is significantly different from 65°C.  
- If p-value ≥ 0.05, **fail to reject** → No evidence to say it differs from 65°C.


# **Independent t-test** ⚔️

## **What is it?**  
Compare the means of two independent groups to see if they differ.

## **Why use it?**  
To test if two separate groups (e.g., two treatments, two classes) have different average values.

## **Real Case (Framed Question)**  
You want to know:  
*"Do students taught with Method A score differently from those taught with Method B on the same test?"*

## **Python Code**

```python
from scipy.stats import ttest_ind

# Example scores
scores_A = [85, 90, 88, 92, 87]
scores_B = [78, 80, 75, 83, 79]

t_stat, p_value = ttest_ind(scores_A, scores_B, equal_var=True)  # or False for Welch’s

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

## **Insights**

- p-value < 0.05: Methods produce significantly different scores.  
- p-value ≥ 0.05: No significant difference between teaching methods.


# **Paired t-test** 🔄

## **What is it?**  
Compare the means of two related groups (e.g., same subjects tested twice).

## **Why use it?**  
To test if a treatment or intervention changed something within the same group.

## **Real Case (Framed Question)**  
You ask:  
*"Did the diet program reduce the weight of participants after 3 months?"*

```python
import numpy as np
from scipy import stats

# Weights of participants before and after diet
before = np.array([75, 80, 85, 90, 95])
after = np.array([74, 78, 83, 88, 94])

# Run paired t-test
t_stat, p_value = stats.ttest_rel(before, after)

print(f"t-statistic: {t_stat:.3f}, p-value: {p_value:.3f}")
```

## **Insights**

- p-value < 0.05: Significant weight loss after diet.  
- p-value ≥ 0.05: No significant change in weight.


# **Z-Test** 🧮

## **What is it?**  
A Z-test is a statistical test to determine if there is a significant difference between a sample mean and a known population mean (or between two sample means) when the population variance is known and/or when the sample size is large (usually n > 30). It’s similar to the one-sample t-test but used when population standard deviation is known.

## **Why use it?**  
To test hypotheses about population means when you know the population variance or have a large sample.  
Faster and simpler than a t-test in such cases.

## **Real Case (Framed Question)**  
A factory produces light bulbs with a claimed average lifespan of 1000 hours. The population standard deviation is known to be 80 hours. You sample 50 bulbs with an average lifespan of 980 hours. You want to know:  
*"Is the average lifespan of bulbs different from the claimed 1000 hours?"*

## **Python Code for One-Sample Z-Test**

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 980
pop_mean = 1000
pop_std = 80
n = 50

# Calculate standard error
std_error = pop_std / np.sqrt(n)

# Calculate z-score
z_score = (sample_mean - pop_mean) / std_error

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}, p-value: {p_value:.3f}")
```

## **Insights**

- If p-value < 0.05: Reject the null hypothesis — the average lifespan is significantly different from 1000 hours.  
- If p-value ≥ 0.05: No significant difference — the lifespan matches the claim.

## **Introduction to Non-Parametric Tests**

Non-parametric tests are statistical methods used when data doesn’t meet the assumptions required for parametric tests, such as normality or equal variances. They are often used for ordinal data or when sample sizes are small.

## **When to Use Non-Parametric Tests**

- Data is not normally distributed  
- Sample size is too small to confirm normality  
- Data is ordinal or ranked  
- When you cannot assume equal variances  

## **Comparison with Parametric Tests**

| Feature             | Parametric Tests                      | Non-Parametric Tests                     |
|---------------------|-------------------------------------|-----------------------------------------|
| Data Type           | Interval or ratio scale              | Ordinal or non-normal interval/ratio    |
| Distribution Assumption | Assumes normal distribution         | No strict distribution assumptions       |
| Sensitivity         | More powerful if assumptions met    | Less powerful but more flexible          |
| Examples            | t-test, ANOVA                       | Mann-Whitney U, Wilcoxon, Kruskal-Wallis|


# **Chi-Square Test (Goodness of Fit & Independence)** 🧮

## **What is it?**  
- **Goodness of Fit:** Tests if observed categorical data matches an expected distribution.  
- **Test of Independence:** Tests if two categorical variables are related or independent.

## **Why use it?**  
When data is categorical (counts or frequencies) and you want to check if distributions differ or variables are associated.

## **Real Case (Framed Question)**  
- **Goodness of Fit:**  
  A dice is rolled 60 times. Are the results uniformly distributed?  
  *"Does the dice behave fairly?"*  
- **Test of Independence:**  
  Do smoking habits (smoker/non-smoker) relate to lung disease status (yes/no)?  
  *"Is smoking associated with lung disease?"*

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table: rows=Smoking status, columns=Lung disease status
# Example data: [[Smoker & Disease, Smoker & No Disease],
#                [Non-Smoker & Disease, Non-Smoker & No Disease]]
data = np.array([[30, 70],
                 [20, 80]])

chi2, p, dof, expected = chi2_contingency(data)
print(f"Chi2 statistic: {chi2:.3f}, p-value: {p:.3f}")
```

## **Insights**

- p-value < 0.05: Variables are associated (dependent).  
- p-value ≥ 0.05: No association (independent).


# **Mann-Whitney U Test** 🆚

## **What is it?**  
A non-parametric test comparing the medians of two independent groups.

## **Why use it?**  
Use when data isn’t normally distributed or is ordinal and you want to compare two groups.

# **Explanation of A/B Testing**

### ✅ What is A/B Testing?
A/B Testing (also known as split testing) is a method of comparing two versions of a product, webpage, or feature to determine which one performs better.  
You show **Version A** to one group and **Version B** to another, then measure performance metrics like **click-through rate**, **conversion rate**, or **engagement**.

### 🧪 Real-Life Example:
Suppose an e-commerce website wants to test two versions of a "Buy Now" button:

- **Version A:** Green button  
- **Version B:** Red button

Half the visitors see the green button, and half see the red. After a few days, the website analyzes which version got more purchases.

### 📊 Key Terminologies:

- **Control Group (A):** The current or default version  
- **Treatment Group (B):** The new or modified version  
- **Conversion Rate:** % of users who completed the desired action  
- **p-value:** Probability that the observed difference is due to chance  
- **Significance Level (α):** Commonly 0.05 (5%)

### 🎯 Steps to Perform A/B Testing:

1. Define the goal (e.g., increase signup rate)  
2. Create two variants (A and B)  
3. Split your audience randomly into two groups  
4. Run the test for a sufficient time to gather data  
5. Perform statistical analysis to check significance  
6. Choose the best version based on results

### A/B Testing in Python
Let’s use a Z-test for proportions to determine if the difference in conversion rates is statistically significant.
### ✨ Scenario:
- **Group A (control):** 1000 users, 200 conversions  
- **Group B (variant):** 980 users, 240 conversions

```python
from statsmodels.stats.proportion import proportions_ztest

# Conversions and sample sizes
conversions = [200, 240]  # A and B conversions
samples = [1000, 980]     # A and B total users

# Perform the z-test
stat, p_value = proportions_ztest(conversions, samples)

# Print results
print("Z-statistic:", stat)
print("P-value:", p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Result: Statistically significant difference ✅")
else:
    print("Result: No statistically significant difference ❌")
```

📌 Interpretation:

✅ If p-value < 0.05, the difference is statistically significant → Choose Version

❌ If p-value ≥ 0.05, there's no strong evidence that version B is better



# ABC Testing Explained

**ABC Testing** is an extension of A/B Testing where you compare three or more versions (A, B, C, etc.) to see which performs best.

## How ABC Testing Works
1. Create three versions:  
   - A (control)  
   - B (variant 1)  
   - C (variant 2)  
2. Randomly split visitors into three groups.  
3. Each group sees one version.  
4. Track a key metric (e.g., clicks, conversions).  
5. Analyze which version performs best statistically.

## Example Scenario  
Test three button colors on a webpage:  
- Version A: Blue button  
- Version B: Green button  
- Version C: Red button  

Show these buttons randomly to visitors and measure click rates to find the best color.


```python
import numpy as np
from scipy import stats

# Data for each version: visitors and clicks
data = {
    'A': {'visitors': 1000, 'clicks': 200},  # 20% conversion
    'B': {'visitors': 1000, 'clicks': 250},  # 25% conversion
    'C': {'visitors': 1000, 'clicks': 230},  # 23% conversion
}

# Calculate and print conversion rate for each version
for version in data:
    clicks = data[version]['clicks']
    visitors = data[version]['visitors']
    conversion_rate = clicks / visitors
    data[version]['conversion_rate'] = conversion_rate
    print(f"Conversion rate for version {version}: {conversion_rate:.2%}")

# Function to do z-test comparing conversion rates of two versions
def z_test(clicks1, visitors1, clicks2, visitors2):
    p1 = clicks1 / visitors1
    p2 = clicks2 / visitors2
    p_combined = (clicks1 + clicks2) / (visitors1 + visitors2)  # pooled rate
    # Standard error for difference between two proportions
    se = np.sqrt(p_combined * (1 - p_combined) * (1/visitors1 + 1/visitors2))
    z = (p1 - p2) / se  # z-score for difference
    p_val = 2 * (1 - stats.norm.cdf(abs(z)))  # p-value (two-tailed test)
    return z, p_val

# Compare pairs of versions and print results
pairs = [('A', 'B'), ('B', 'C'), ('A', 'C')]

for v1, v2 in pairs:
    z, p = z_test(data[v1]['clicks'], data[v1]['visitors'], data[v2]['clicks'], data[v2]['visitors'])
    print(f"Compare {v1} vs {v2} -> Z = {z:.4f}, P-value = {p:.4f}")
    if p < 0.05:
        print(f"Result: The difference between {v1} and {v2} is statistically significant.\n")
    else:
        print(f"Result: No significant difference between {v1} and {v2}.\n")
```

# Insights

- Version B has a significantly higher conversion rate than Version A — so B is likely better than A.

- There is no strong evidence that Version B performs differently than Version C.

- Version C performs better than A, but the difference is not statistically significant at 5% level, though it’s somewhat close.

- You might consider choosing Version B as the winner since it has the highest conversion rate and is statistically better than A.

- For more certainty about C vs A, you could increase your sample size or run the test longer.

|------------|---------|---------|------------------------------------------------|
| A vs B     | ~ -3.12 | ~0.0018 | Significant difference: B performs better than A |
| B vs C     | ~ 1.10  | ~0.2711 | No significant difference between B and C      |
| A vs C     | ~ -1.70 | ~0.0890 | No significant difference at 0.05 level, but close |


# What is Multi-Variant Testing?

**Multi-Variant Testing (MVT)** is a testing method used to evaluate multiple variables and their combinations simultaneously on a webpage or product feature to see which combination performs best.

Unlike A/B Testing (which compares two versions) or ABC Testing (which compares three versions), MVT can test several elements (like headlines, images, button colors) at once, by mixing different variations of each element.

---

## How Multi-Variant Testing Works

Suppose you want to test 2 different headlines (H1, H2) and 3 different button colors (Blue, Green, Red).

MVT will create all possible combinations:

- H1 + Blue  
- H1 + Green  
- H1 + Red  
- H2 + Blue  
- H2 + Green  
- H2 + Red  

You split your traffic randomly into these 6 groups.

Track conversion rates or other metrics for each combination.

Analyze which combination (or which individual elements) leads to the best performance.

---

## Why Use Multi-Variant Testing?

- It helps understand interactions between different elements.  
- More efficient than testing one element at a time.  
- Finds the best combination rather than just the best individual change.

# What is Two-Way ANOVA?

**Two-Way ANOVA** (Analysis of Variance) is a statistical test used to determine the effect of two independent categorical variables (factors) on a continuous dependent variable, and also whether there is an interaction effect between the two factors.

---

## When to use Two-Way ANOVA?

- You have **two factors** (e.g., A and B).
- You want to see if **each factor independently affects the outcome**.
- You also want to check if there’s an **interaction effect** (i.e., whether the effect of one factor depends on the level of the other).

## Example Scenario

Suppose you want to test the effect of:

- **Factor 1:** Type of fertilizer (Fertilizer A and Fertilizer B)
- **Factor 2:** Water amount (Low and High)

on the growth of plants (height in cm).

You measure plant heights under all combinations:

| Fertilizer | Water | Heights (cm)          |
|------------|-------|-----------------------|
| A          | Low   | 20, 21, 19, 22        |
| A          | High  | 25, 26, 27, 28        |
| B          | Low   | 18, 17, 20, 19        |
| B          | High  | 23, 24, 22, 25        |

You want to know:

- Does fertilizer type affect plant growth?
- Does water amount affect plant growth?
- Is there an interaction between fertilizer and water?

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create the dataset
data = {
    'Fertilizer': ['A']*8 + ['B']*8,
    'Water': ['Low']*4 + ['High']*4 + ['Low']*4 + ['High']*4,
    'Height': [20, 21, 19, 22, 25, 26, 27, 28, 18, 17, 20, 19, 23, 24, 22, 25]
}

df = pd.DataFrame(data)

# Fit the two-way ANOVA model with interaction
model = ols('Height ~ C(Fertilizer) + C(Water) + C(Fertilizer):C(Water)', data=df).fit()

# Perform ANOVA table
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

### Insights
Fertilizer and Water both have significant effects (p < 0.05).

Interaction is not significant (p > 0.05), so no evidence that the effect of fertilizer depends on water leve

## ⚠️ Shortcomings

🎯 Assumptions may not hold true in real-world data  
🔢 Small sample sizes can reduce reliability  
🧪 Risk of experimental bias without proper controls  
⏱️ Time-consuming and resource-intensive  
📉 P-values can be misunderstood or misused  
🔄 Difficult to control all variables outside lab settings  
🧠 Results may not generalize to other contexts


## ✅ Best Practices

🎯 Clearly define the research question or hypothesis before starting.  
🧪 Use control groups and randomization to reduce bias.  
🔢 Ensure a large enough sample size for reliable results.  
📏 Choose appropriate statistical tests based on data type and distribution.  
📊 Visualize data to uncover patterns and communicate findings.  
🔍 Always validate assumptions before applying parametric tests.  
🔁 Replicate experiments to confirm results and ensure reliability.  
📚 Document all procedures, variables, and changes during the experiment.  
🧠 Interpret results with critical thinking, not just p-values.  
🔐 Follow ethical guidelines and maintain data privacy throughout.
