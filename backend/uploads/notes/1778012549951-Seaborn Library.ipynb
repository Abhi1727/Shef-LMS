# The primary difference between Matplotlib and Seaborn is their level of abstraction. Matplotlib is a low-level library that acts like a blank canvas, giving you absolute control over every pixel of a plot but requiring more code. Seaborn is a high-level wrapper built on top of Matplotlib, designed to create "pretty" statistical charts with minimal code.

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# seaborn
import seaborn as sns
```

```python
# dataset
df = pd.read_csv('/content/tip.csv')
```

```python
df.head()
```

```python
# rename the sex column
df = df.rename(columns = {'sex' : 'gender'})
```

```python
df.head()
```

# Common Plots and Charts

```python
# scatterplot -- total bill vs tips
sns.scatterplot(data = df, x = 'total_bill', y = 'tip')
plt.title('Total Bill vs Tip', fontsize = 15)
```

```python
# scatterplot -- total bill vs tips
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', color = 'red')
plt.title('Total Bill vs Tip', fontsize = 15)
plt.grid(True)
```

# **1. Strip Plot**
- A strip plot (or individual value plot) is a scatter plot that displays the distribution of continuous data points along a single axis, often grouped by a categorical variable. It is excellent for visualizing the spread, density, and outliers in small datasets, frequently serving as a complement to box or violin plots to show raw data.

```python
# display the distribution of total bill on day wise
sns.stripplot(data = df, x = 'day', y = 'total_bill')
```

# add the color palette for better visua;ization

```python
# ignore the future warnings during the execution of code
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.stripplot(data = df, x = 'day', y = 'total_bill', palette = 'bright')
plt.title('Day wise bill Distribution', fontsize = 15, color = 'blue')
```

```python
# jitter -- shrink the distribution of data points
sns.stripplot(data = df, x = 'day', y = 'total_bill', palette = 'muted', jitter = False)
plt.title('Day wise bill Distribution', fontsize = 15, color = 'blue')
```

# hue parameter
The hue parameter in Seaborn is a powerful semantic mapping tool used to color data points based on a specific variable. By assigning a variable to hue, you add a third dimension to your visualization, allowing you to see how relationships between your main variables change across different categories or numerical levels.


```python
sns.stripplot(data = df, x = 'day', y = 'total_bill',
              hue = 'gender', palette = 'bright')
plt.title('Day wise bill Distribution', fontsize = 15, color = 'blue')
```

# **2. Swarm plot**
A swarm plot is a categorical scatterplot that displays individual data points without overlapping, adjusting them along the categorical axis to reveal the underlying distribution, similar to a violin plot. Ideal for small to moderate datasets, it offers a clear view of data density, frequency, and clusters, often used alongside box or violin plots.

```python
sns.swarmplot(x = df['day'], y = df['total_bill'],
              palette = 'bright', hue = df['gender'])
```

# Boxplot

catplot --- category

```python
sns.catplot(data = df, x = 'day', y = 'tip', kind = 'box', palette = 'bright')
```

create a boxplot using catplot applying the hue parameter -- group by gender.

```python
sns.catplot(data = df, x = 'day', y = 'tip', kind = 'box',
            hue= 'gender', palette = 'bright')
```

# 4. violin plot
A violin plot is a data visualization tool that combines a box plot with a kernel density estimate (KDE) to show the density and distribution of numerical data across categories. The wider sections represent higher data density, while narrower sections show lower density. It is superior to box plots for showing multiple peaks (multimodality) and complex distributions.

```python
# voilin = (boxplot + KDE)
sns.violinplot(data = df, x = 'day', y = 'total_bill',
               palette = 'bright')
plt.title('Violin Plot (boxplot + KDE)', fontsize = 14, color = 'blue')
```

```python
sns.violinplot(data = df, x = 'day', y = 'total_bill', hue = 'gender',
               palette = 'bright')
plt.title('Violin Plot (boxplot + KDE)', fontsize = 14, color = 'blue')
```

```python
sns.violinplot(data = df, x = 'day', y = 'total_bill', hue = 'gender',
               color = 'cyan', split = True)
plt.title('Violin Plot (boxplot + KDE)', fontsize = 14, color = 'blue')
```
