# Dictionary Built-in Functions Summary

| Function/Method | Description | Example |
|---|---|---|
| `dict()` | Creates a dictionary. | `info = {'name':'Ridaa','Age':25,'place':'Bangalore'}` |
| `.keys()` | Returns a view object that displays a list of all the keys in the dictionary. | `info.keys()` |
| `.values()` | Returns a view object that displays a list of all the values in the dictionary. | `info.values()` |
| `.items()` | Returns a view object that displays a list of a dictionary's key-value tuple pairs. | `info.items()` |
| `dict[key] = value` | Adds a new key-value pair or updates the value for an existing key. | `info['company']='IBM'` |
| `.get(key)` | Returns the value for the specified key if it is in the dictionary. | `info.get('Age')` |
| `.update(dict2)` | Updates the dictionary with the key-value pairs from another dictionary or from an iterable of key-value pairs. | `info.update({'name':'Zara'})` |
| `.setdefault(key, value)` | Returns the value of the key if it is in the dictionary, otherwise inserts the key with the specified value and returns the value. | `info.setdefault('salary',60000)` |
| `.pop(key)` | Removes the item with the specified key and returns its value. | `info.pop('exp')` |
| `.popitem()` | Removes the last inserted key-value pair and returns it as a tuple. | `info.popitem()` |
| `.clear()` | Removes all the elements from the dictionary. | `info.clear()` |
| `del dict` | Deletes the entire dictionary. | `del info` |
| `.copy()` | Returns a shallow copy of the dictionary. | `B = A.copy()` |
