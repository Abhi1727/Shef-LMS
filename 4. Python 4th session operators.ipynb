# Logical operators
Logical operators are used to check multiple conditions.
* and
* or
* not

```python
# and : If all the conditions are TRue it returns True , otherwise false
'''
TT : T
FT : F
TF : F
FF : F '''
print((5 < 10) and (100>20))
# True and True : True
```

```python
(2<100) and (100>2000)
# True and False : False
```

```python
(210<100) and (100>20)
# False and True : False
```

```python
(3>300) and (4>400)
# False and False
```

```python
# Or : It gives true if either of the condition is True , otherwise false
'''
T T : T
T F : T
F T : T
F F : F
'''

print((2<100) or (3<300))
# True or True
```

```python
(300>3) or (20>400)
# True or False
```

```python
(300<3) or (200>400)
# False  or False
```

```python
# not : If condition is True it gives False and vice versa
not(100 > 10)
```

```python
not(10>100)
```
