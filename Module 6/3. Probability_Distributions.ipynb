# 🎯 What is Binomial Distribution?

The **Binomial Distribution** is a **discrete probability distribution** that models the number of **successes** in a **fixed number of trials**, where each trial has:

- ✅ Only **two outcomes**: Success or Failure  
- 🎯 The **same probability** of success  
- 🔄 Each trial is **independent**

Snip20250422_53.png

## Why It’s Important in Data Science & Machine Learning

Binomial distribution plays a critical role in:

- 🌐 **Modeling conversion rates** (click-throughs, purchases)
- 🧪 **A/B testing** — estimating likelihood of user actions
- 🤖 **Logistic Regression** uses binomial logic to predict binary outcomes
- 📬 **Email campaigns**, **fraud detection**, **medical tests** — all involve **binary outcomes**!


## Real Case Scenario: Email Campaign Click Prediction

👩‍💻 You're a **data analyst** at an **online education platform**.

You send a **newsletter to 500 users**, and from past data, you know:

- **Click-through rate (CTR)** = 20% or 0.2

### ❓ Business Question:
> What is the probability that **exactly 120 users** will click on the email?

We solve this using the **Binomial Distribution** in Python.


#### Python Code Example

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
n = 500     # total emails sent
p = 0.2     # click-through rate (probability of success)
k = 120     # exact number of clicks we are checking for

# Probability of exactly 120 clicks
prob = binom.pmf(k, n, p)
print(f"🔍 Probability of exactly 120 users clicking the email: {prob:.5f}")
```

# 📈 Poisson Distribution

The **Poisson Distribution** models the **number of times an event occurs** in a **fixed interval of time or space**, given a known **average rate (λ)**, and assuming events happen **independently**.

In simple terms, it answers questions like:

- "How many calls will come into the call center per hour?"
- "How many users will visit the site in a minute?"

##  Real-World Scenario: API Error Monitoring

### 📌 Scenario:

You're a **backend engineer** tracking the number of **error logs** your server gets per minute.

From past data, you know your app gets an average of **4 errors per minute** (λ = 4).

Now you want to know:

> ❓ "What’s the probability that **7 errors** will occur in a minute?"

#### Python Code Example

```python
from scipy.stats import poisson

# Average rate of errors per minute
λ = 4  # (lambda)
k = 7  # we want to find P(X=7)

# Probability of exactly 7 errors in a minute
prob = poisson.pmf(k, λ)
print(f"🔍 Probability of getting exactly 7 errors in a minute: {prob:.4f}")
```

##  Real-World Scenario: Ad Campaign Success

### 📌 Scenario:

You're running a **Facebook ad campaign**.  
Based on past data, you know:

- Probability a user **clicks an ad** = 0.1 (10%)

You want to answer:

> ❓ "What’s the probability that the **first click** happens on the **5th user**?"

```python
from scipy.stats import geom

# Parameters
p = 0.1   # probability of success
k = 5     # trial where success happens

# Probability first success occurs on 5th trial
prob = geom.pmf(k, p)
print(f"🔍 Probability of first click on the 5th user: {prob:.4f}")
```

The **Bernoulli Distribution** models a **single trial (or experiment)** that has exactly **two possible outcomes**:

- **Success (1)**
- **Failure (0)**

### It answers:
> “What’s the chance of success (or failure) in one attempt?”

| 💡 Basis of Other Distributions | Binomial = sum of Bernoulli trials            |

➡️ Basically, **whenever your outcome is binary**, **Bernoulli** is the go-to model.

## Real-World Scenario: Ad Click

### 📌 Scenario:

You show a user a **digital ad**.  
Based on past data, you know:

- Probability a user **clicks** = **0.25** (25%)

You want to answer:

> ❓ “What’s the probability that this user **clicks (1)**?”  
> ❓ “What’s the probability that this user **doesn’t click (0)**?”

```python
from scipy.stats import bernoulli

# Probability of success
p = 0.25  # 25% chance the user clicks

# Probability of click (success = 1)
click = bernoulli.pmf(1, p)

# Probability of no click (failure = 0)
no_click = bernoulli.pmf(0, p)

print(f"✅ Probability of click: {click:.2f}")
print(f"❌ Probability of no click: {no_click:.2f}")
```

# 🔷 Uniform Distribution

The **Uniform Distribution** is a **continuous probability distribution** where **every value** within a specific interval \([a, b]\) is **equally likely** to occur.

### 📌 Key Characteristics:
- The probability is **constant** across the interval.
- Used when **no value** is more likely than another within the range.
- The shape of the distribution is a **rectangle** (also called a rectangular distribution).

## 🎯 Real-World Scenario (Framed as a Question)

You're a **data analyst** working on a **video marketing campaign**.

A **60-second video** is shown to users, and you **randomly choose a time** to display a popup ad.

### ❓ Question:

> **What’s the probability that the ad appears between the 20th and 30th second?**


```python
from scipy.stats import uniform

# Parameters
a = 0       # start of interval
b = 60      # end of interval

# Calculate probability between 20s and 30s
prob = uniform.cdf(30, loc=a, scale=b - a) - uniform.cdf(20, loc=a, scale=b - a)
print(f"📢 Probability the ad appears between 20s and 30s: {prob:.2f}")
```

# 📈 Normal Distribution (Gaussian)

The **Normal Distribution** is a **continuous probability distribution** where values are **symmetrically distributed around a mean (μ)**, forming the famous **bell-shaped curve**.

It's one of the most important and widely used distributions in statistics and machine learning.

## 🎯 Real-World Scenario (Framed as a Question)

You're analyzing the **exam scores** of students in a **large national test**.  
The average score is **70**, with a standard deviation of **10**.

---

### ❓ Question:

> **What’s the probability that a student scores between 60 and 80?**

```python
from scipy.stats import norm

mu = 70        # average score
sigma = 10     # standard deviation

# Calculate probability between 60 and 80
prob = norm.cdf(80, loc=mu, scale=sigma) - norm.cdf(60, loc=mu, scale=sigma)
print(f"📚 Probability that a student scores between 60 and 80: {prob:.2f}")
```

There’s about a 68% chance that a randomly selected student scores between 60 and 80.

# 📏 Empirical Rule (68-95-99.7 Rule)
The **Empirical Rule** describes how data is distributed in a **Normal Distribution (bell curve)**.  
It tells us how much of the data falls within **1, 2, and 3 standard deviations** from the mean.


## 📊 The 68-95-99.7 Rule:

| Range                          | % of Data Included |
|-------------------------------|---------------------|
| μ ± 1σ (1 standard deviation) | ~68%               |
| μ ± 2σ (2 standard deviations) | ~95%               |
| μ ± 3σ (3 standard deviations) | ~99.7%             |

Where:
- μ = Mean (center of the distribution)  
- σ  = Standard deviation (spread of the data)

# What is the Standard Normal Distribution?

The **Standard Normal Distribution** is a special case of the **normal distribution** (also called the **Gaussian distribution**).

It has:

- **Mean (μ) = 0**
- **Standard Deviation (σ) = 1**

In short:

> It's a normal (bell-shaped) curve that is perfectly centered at 0 and scaled so that the spread (standard deviation) is exactly 1.



## 🎯 Key Characteristics:

- Symmetrical around the mean (**0**).
- Bell-shaped curve.
- Total area under the curve = **1** (representing 100% probability).
- Most data falls within:
  - **±1 standard deviation** ➔ about **68%** of the data
  - **±2 standard deviations** ➔ about **95%**
  - **±3 standard deviations** ➔ about **99.7%**

(👉 This is called the **Empirical Rule**.)



## 🧠 Why is it important?

- **Simplifies probability calculations:**
  - Instead of using different normal distributions, we standardize any normal random variable using a **Z-score**.
- **Basis for Z-tables:**
  - Z-tables tell us cumulative probabilities associated with different z-scores in a standard normal distribution.
- **Foundation for Hypothesis Testing:**
  - Many statistical methods like calculating **p-values**, **confidence intervals**, and **hypothesis testing** use the standard normal distribution.

# ⚡ Exponential Distribution

## ✅ Definition:

The **Exponential Distribution** models the **time between consecutive events** in a process where events happen:

- **Continuously**
- **Independently**
- At a **constant average rate (λ)**



image.png

## 🎯 Real-World Scenario

You're monitoring an **API server**.  
On average, it receives **1 request every 6 seconds**.

##### ❓ Question:

> **What is the probability that the next request comes after 10 seconds?**


```python
from scipy.stats import expon

# Given: average time between requests = 6s → λ = 1/6
lambda_rate = 1 / 6
scale = 1 / lambda_rate

# Probability next request comes after 10s
prob = expon.sf(10, scale=scale)  # sf = 1 - cdf
print(f"⏳ Probability of waiting more than 10 seconds: {prob:.2f}")
```

There is about a 19% chance the next request takes longer than 10 seconds to arrive.

# Best Practices for Probability Distributions:

1. **Understand the real-world scenario** you're modeling before applying any distribution (PMF, CDF, etc.) to ensure you choose the right distribution. 🧠

2. **Verify trial independence**: For discrete distributions like the Binomial, ensure that trials are independent, as this is a key assumption. 🔄

3. **Use the CDF**: The CDF is useful to calculate the probability of a random variable being less than or equal to a specific value, particularly for discrete data. 📉

4. **Calculate Expected Value and Variance**: Always calculate both the expected value and variance to understand the central tendency and spread of the data. 📏📍

5. **Validate the PMF**: Ensure that the sum of the PMF probabilities equals 1 and that the values are non-negative. ✔️



## For Continuous Distributions:

6. **Understand the PDF**: For continuous distributions, the Probability Density Function (PDF) tells you the likelihood of a variable falling within a specific range. 📈

7. **Use the CDF for continuous distributions**: The CDF helps you understand the cumulative probability of a value being less than or equal to a certain value. 📊

8. **Verify data assumptions**: Ensure that the data fits the assumptions of the distribution being used (e.g., Normal, Exponential). Also, correctly estimate parameters like mean and variance for continuous distributions. 📊🧮


# Shortcomings of Probability Distributions:

1. **Limited to Assumed Conditions**: 🔒 Many distributions, such as the Normal or Binomial, assume certain conditions (e.g., independence, constant rate) which may not always be met in real-world data. This can lead to inaccurate results if the assumptions are violated.

2. **Inaccurate for Complex Data**: 📊 Some distributions, like the Poisson or Binomial, may oversimplify complex data patterns. They may not capture nuances or non-linear relationships, making them unsuitable for more intricate real-world problems.

3. **Sensitive to Outliers**: 🚨 Distributions like the Normal Distribution are sensitive to outliers. Outliers can distort the mean and variance, leading to misleading conclusions if not properly handled.

4. **No Exact Probability for Continuous Data**: 📉 For continuous distributions, the probability of a random variable taking an exact value is zero. This can be counterintuitive, especially when comparing to discrete data where exact probabilities can be meaningful.

5. **Challenges in Parameter Estimation**: 🧮 Estimating distribution parameters (like mean, variance, etc.) can be challenging, especially for small datasets or when data doesn't fit a known distribution.

6. **Overfitting Risk**: ⚠️ When applying distributions to fit data, there’s a risk of overfitting, especially with more flexible models. This can lead to poor generalization to new, unseen data.

7. **Computational Complexity**: 💻 Some distributions require complex mathematical computations, which can be time-consuming and computationally expensive, especially for large datasets.

8. **Misapplication of Models**: ❌ Incorrectly applying the wrong distribution (e.g., using a Normal Distribution when the data is skewed) can lead to inaccurate predictions and poor decision-making.

2. The time between arrivals at a certain store follows an exponential distribution with a mean of 8 minutes. What is the probability that the time between two consecutive arrivals will be less than 5 minutes?




# Task 2

1. A fair six-sided die is rolled two times. What is the probability of getting a sum of 7?
2. Suppose the heights of a group of people are normally distributed with a mean of 68 inches and a standard deviation of 3 inches. What is the probability that a randomly selected person from this group is shorter than 70 inches tall?
