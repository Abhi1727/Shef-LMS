## 🧠 Three Pillars of Descriptive Statistics

Descriptive statistics is typically broken down into **three types**:

### 1️⃣ Central Tendency – *“What’s Typical?”*

These measures tell you where the **center** of the data lies.

| Measure | Meaning                                   | Example                        |
|--------|-------------------------------------------|-------------------------------|
| Mean   | Arithmetic average                         | Average salary in a company   |
| Median | Middle value (less affected by outliers)   | Middle score in an exam       |
| Mode   | Most frequently occurring value            | Most common shoe size         |


### 2️⃣ Dispersion – *“How Much Does It Vary?”*

Dispersion shows the **spread** of your data. It tells you how much values differ from the center.

| Measure            | Meaning                                               | Use Case                             |
|--------------------|-------------------------------------------------------|--------------------------------------|
| Range              | Max - Min                                             | Price range of smartphones           |
| Variance           | Average of squared differences from the mean          | Assessing risk in stock prices       |
| Standard Deviation | Square root of variance (more interpretable)          | Consistency of delivery times        |
| Interquartile Range (IQR) | Spread of the middle 50% of the data             | Filtering out extreme income values  |

📌 **Note:** A low standard deviation = values are tightly clustered.  
A high standard deviation = values are spread out.


### 3️⃣ Shape (Distribution) – *“How Does the Data Behave?”*

This tells you about the **form** or **pattern** of the data when plotted.

| Feature   | Description                                        | Example                                      |
|-----------|----------------------------------------------------|----------------------------------------------|
| Skewness  | Asymmetry in the distribution                      | Income data is often **right-skewed**        |
| Kurtosis  | Sharpness or flatness of the peak                  | Exam scores with many average marks = flat   |

#### Visual Summary:
- **Normal Distribution:** Symmetrical bell curve
- **Positive Skew:** Tail on the right
- **Negative Skew:** Tail on the left
- **Leptokurtic:** Tall and thin
- **Platykurtic:** Flat and wide

# 📌 **What is Measure of Central Tendency?**

**Measure of Central Tendency** refers to statistical values that describe the **center point** or **typical value** of a dataset.

##### In simple words:
> It helps answer the question – *"What is the average or most typical value in the data?"*

These measures reduce a large dataset to a **single representative value** that summarizes the data.


#### 📊 Why is it Useful?

##📌**What is Mean?**
The **mean**, also known as the **arithmetic average**, is one of the most commonly used measures of **central tendency**.
It represents the average value of a dataset.

It gives you a **single number** that represents the **center point** of your dataset.

**In simple terms:**
> Add up all the values and divide by how many values you have.

✅ Simple to calculate and understand

✅ Uses all values in the dataset

✅ Ideal when data is symmetrically distributed

❗️Not ideal if there are outliers (extremely high or low values)



#📌 **What is Median?**
Median is the middle value of a dataset when it is sorted in order.

It represents the point that divides the data into two equal halves — 50% of the values lie below and 50% above.

**🧮 How to Find Median:**

🔹 **If n is odd:**

The median is the middle value.

🔹 **If n is even:**

The median is the average of the two middle values.

#### **Python code**

```python
import numpy as np

# Odd number of values
data_odd = [12, 5, 8, 20, 15]
median_odd = np.median(data_odd)
print("Median (odd n):", median_odd)

# Even number of values
data_even = [10, 20, 30, 40, 50, 60]
median_even = np.median(data_even)
print("Median (even n):", median_even)
```

###🎯 **Why Use Median?**
✅ Not affected by outliers or extreme values

✅ Better than mean for skewed distributions

✅ Useful when data is ordinal (e.g., survey ratings)

# 📌 **What is Mode?**

The **mode** is the value that appears **most frequently** in a dataset.

It tells you the **most common** or **popular** value — a great indicator of trends or majority preference.


## 🧮 How to Find Mode

1. Count how many times each value appears in the dataset.
2. The value(s) with the **highest frequency** is the **mode**.


## 🧠 Special Cases

| Case Type       | Description                                      |
|------------------|--------------------------------------------------|
| **Unimodal**     | Only one value appears most frequently            |
| **Bimodal**      | Two values have the same highest frequency        |
| **Multimodal**   | More than two values are equally frequent         |
| **No Mode**      | All values appear the same number of times        |


## ❓ Sample Questions

### ✅ Case 1: One Mode (Unimodal)

**Q:** Find the mode of `[2, 4, 4, 6, 8]`  
**A:** `4` appears most frequently  
**→ Mode = 4**


### ✅ Case 2: Two Modes (Bimodal)

**Q:** Find the mode of `[1, 2, 2, 3, 3, 4]`  
**A:** Both `2` and `3` appear **twice**  
**→ Mode = 2 and 3** (Bimodal)

### ✅ Case 3: No Mode

**Q:** Find the mode of `[1, 2, 3, 4, 5]`  
**A:** All values appear **once**  
**→ No Mode**


## 📊 When is Mode Useful?

- **Categorical Data**: E.g., most common shirt size (M, L, XL)
- **Skewed Distributions**: Gives a better “typical value” than mean
- **Finding Trends**: In marketing, e-commerce, surveys, etc.


## ✅ Quick Tip

> Use **mode** when you're looking for **"what’s most popular?"**  
> It's especially useful for **non-numerical** or **repetitive** data.



### **Python code**

```python
from scipy import stats

# One mode
data1 = [2, 4, 4, 6, 8]
mode1 = stats.mode(data1, keepdims=False)
print("Mode (one mode):", mode1.mode)

# Two modes
import pandas as pd
data2 = [1, 2, 2, 3, 3, 4]
modes = pd.Series(data2).mode()
print("All modes:", list(modes))

```

# 📊 **Measures of Dispersion**
**"How far do your data points wander from the center?"**


## 📌 What is Dispersion?

**Dispersion** refers to the **spread** or **variability** of data values in a dataset.  
It answers the question:  
> *"Are my data points tightly packed or widely scattered?"*

While measures like **mean** and **median** tell you about the **center**,  
**dispersion** tells you how **diverse or consistent** the data is around that center.


## 🔧 Why is Dispersion Important?

- Helps detect **inconsistencies** or **variability**
- Identifies **outliers** or unusual values
- Supports **risk assessment** in areas like finance
- Completes your **EDA story** — central tendency + spread


## 🧠 Key Measures of Dispersion

| Measure            | Description                                      | When to Use                                 |
|--------------------|--------------------------------------------------|---------------------------------------------|
| **Range**          | Difference between max and min values            | Quick check of spread                       |
| **Interquartile Range (IQR)** | Spread of the middle 50% of data     | Outlier detection, box plots                |
| **Variance**        | Average of squared deviations from the mean      | Foundation for standard deviation           |
| **Standard Deviation (SD)** | Square root of variance; real-world spread | Most commonly used, especially in normal data |

---


# 📌 What is Range?

The **Range** is the **simplest measure of dispersion**.  
It tells us how **spread out** the values in a dataset are by showing the **difference between the largest and smallest** values.

In other words, **Range shows the total span** of the data.


#### ✅ How to Calculate Range

Just follow this formula:

Range = Maximum Value - Minimum Value

#### 🧮 Example

Let's say you have the following exam scores:  
`45, 50, 67, 72, 89`

- **Maximum** = 89  
- **Minimum** = 45  

 Range = 89 - 45 = 44


So, the **Range is 44**, meaning there’s a 44-point spread between the lowest and highest scores.


## 🎯 When is Range Useful?

- When you need a **quick snapshot of variability**
- To **compare data spreads** between groups
- In **quality control** to check variation in product dimensions or timings


## ⚠️ Limitations

- **Sensitive to outliers**: One unusually high or low value can drastically change the range.
- Doesn’t give any information about how the **rest of the data is distributed**


## 💡 Tip

> Use **Range** for a quick glance,  
> but pair it with other measures (like **Standard Deviation** or **IQR**) for deeper insights.


```python
# Python code to find range
data = [45, 50, 67, 72, 89]

range_value = max(data) - min(data)

print("Range:", range_value)
```

# 📌 **What is Variance?**
Variance is a way to measure how spread out the numbers in a dataset are.
It tells us how far each number is from the mean (average), and it does this by calculating the average of the squared differences from the mean.

**In short, variance answers the question:**

"On average, how much do the values differ from the mean?"

**✅ How to Calculate Variance:**
1. Find the mean of the data.
2. Subtract the mean from each data point to get the deviation.
3. Square each deviation to make all values positive.
4. Find the average of these squared deviations.

**📌 Variance Formula**

Variance measures the average of the squared deviations from the mean.

1. Use population variance when you have the entire dataset.
2. Use sample variance when you're working with a subset (sample) of the data — dividing by n−1 corrects bias in estimation.

### **📌 Scenario: Supermarket Wait Time Comparison**
You are analyzing customer wait times (in minutes) at two supermarkets:
Supermarket A and Supermarket B.

➡️ Understand which supermarket provides more consistent service — not just faster on average.

#### **Python code**

```python
# Method 1
import statistics

# Data
supermarket_A = [4, 5, 6, 5, 5]
supermarket_B = [1, 7, 2, 10, 0]

# Variance (Population)
var_A = statistics.pvariance(supermarket_A)
var_B = statistics.pvariance(supermarket_B)

# Output
print("Supermarket A - Variance:", var_A)
print("Supermarket B - Variance:", var_B)
```

```python
# Method 2
def calculate_variance(data):
    mean = sum(data) / len(data)
    squared_diffs = [(x - mean) ** 2 for x in data]
    return sum(squared_diffs) / len(data)

# Data
supermarket_A = [4, 5, 6, 5, 5]
supermarket_B = [1, 7, 2, 10, 0]

# Variance
var_A = calculate_variance(supermarket_A)
var_B = calculate_variance(supermarket_B)

print("Supermarket A - Variance:", var_A)
print("Supermarket B - Variance:", var_B)
```

✅ **Supermarket B has:**
- A lower average wait time (**4 mins**)
- But **highly inconsistent service** (high variance of **14.8**)

#### 🎯 **Final Conclusion:**

> **Supermarket A** provides more **consistent and reliable service**,

# 📌 **What is Standard Deviation?**

**Standard deviation** is the **square root of the variance**.  
It tells you, on average, how much the data values differ from the **mean**, but in the **same units** as the data (like minutes, seconds, etc.).

It’s one of the most widely used tools to understand **consistency** and **variability** in a dataset.

#### 🚆 **Scenario: Comparing Train A and Train B Wait Times**

You're analyzing commuter data for two trains and want to understand which one arrives **more consistently** — not just faster.

Even if both trains have the same average wait time, the one with a **lower standard deviation** is more reliable (less variation in arrival times).

```python
import statistics
import math

# Wait times
train_A = [5, 6, 5, 5, 4]
train_B = [2, 10, 3, 9, 1]

# Variance (population)
var_A = statistics.pvariance(train_A)
var_B = statistics.pvariance(train_B)

# Standard Deviation
std_A = math.sqrt(var_A)
std_B = math.sqrt(var_B)

print("Train A - Standard Deviation:", std_A)
print("Train B - Standard Deviation:", std_B)
```

## 🧠 Key Insights:

# 📊 **Quartiles and Interquartile Range (IQR)**

## 📌 What are Quartiles?

**Quartiles** divide a sorted dataset into **four equal parts**, each containing **25% of the data**.

| Quartile | Description                    |
|----------|--------------------------------|
| Q1       | First Quartile (25th percentile) — median of the lower half |
| Q2       | Second Quartile (50th percentile) — also called the median   |
| Q3       | Third Quartile (75th percentile) — median of the upper half  |


## 🔢 Example:  
Given this sorted dataset:  
`[5, 7, 8, 10, 12, 14, 15, 18, 20]`

- **Q2 (Median)** = 12  
- **Lower half** = [5, 7, 8, 10] → Q1 = (7 + 8) / 2 = **7.5**  
- **Upper half** = [14, 15, 18, 20] → Q3 = (15 + 18) / 2 = **16.5**


## 📐 Interquartile Range (IQR)

The **IQR** is the difference between the **third** and **first** quartiles:

IQR = Q3 - Q1

It shows the **spread of the middle 50%** of the data and is **less affected by outliers** than the range.

**Using the example above:**

IQR = 16.5 - 7.5 = 9

## 🎯 Why Use IQR?

- 🛡️ **Robust against outliers** (unlike the range)
- 🔍 Useful for detecting **data concentration**
- 📦 Commonly used in **box plots**
- 🚦 Helps in **outlier detection**  
  Any value below `Q1 - 1.5 × IQR` or above `Q3 + 1.5 × IQR` is often considered an **outlier**


```python
import numpy as np

data = [5, 7, 8, 10, 12, 14, 15, 18, 20]

Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
```

✅ **Final Insight**

IQR gives you a clearer picture of data spread and stability, especially when your dataset contains outliers or skewed values.

# 📌 What are Percentiles?

**Percentiles** are values that divide a dataset into **100 equal parts**.  
Each percentile represents a point **below which a certain percentage of the data falls**.

### 📍 Examples:
- The **25th percentile** means 25% of the values are **less than or equal** to that number.
- The **90th percentile** means 90% of the values are **below or equal** to that value.

## ✅ Why are Percentiles Important?

- **Understanding Distribution**:  
  Percentiles help describe how data is **spread out** and where **most values lie**.

- **Comparing Scores**:  
  Useful in exams, surveys, and test results.  
  *Example:* If your test score is in the **85th percentile**, you scored **better than 85%** of test-takers.

- **Outlier Detection**:  
  Percentiles define **thresholds** for what’s considered *"normal"* vs *"extreme"*.

- **Making Data-Driven Decisions**:  
  Businesses and analysts use percentiles to define **cut-off points** in:
  - Customer behavior  
  - Pricing tiers  
  - Product performance  
  - Employee evaluations


## 🧮 Example:

Let's say we have these **sorted scores**:

`[40, 45, 50, 55, 60, 65, 70, 75, 80, 85]`

- **50th percentile (P50)** = **Median** = (60 + 65) / 2 = **62.5**  
- **25th percentile (P25)** = Position 2.5 → Value between **45 and 50**  
- **75th percentile (P75)** = Position 7.5 → Value between **70 and 75**


## 📊 Percentiles vs Quartiles

| Term        | Meaning                          |
|-------------|----------------------------------|
| Percentile  | Divides data into **100** parts  |
| Quartile    | Divides data into **4** parts    |
| Decile      | Divides data into **10** parts   |

### 🔄 Mapping Quartiles to Percentiles:
- **Q1 = 25th percentile**
- **Q2 = 50th percentile (Median)**
- **Q3 = 75th percentile**



```python
import numpy as np

data = [40, 45, 50, 55, 60, 65, 70, 75, 80, 85]

p25 = np.percentile(data, 25)
p50 = np.percentile(data, 50)
p75 = np.percentile(data, 75)

print("25th Percentile (Q1):", p25)
print("50th Percentile (Median):", p50)
print("75th Percentile (Q3):", p75)
```

🎯 **Final Thought:**

Percentiles help you position a value in a crowd — giving clear context for how high or low it stands compared to others.

## **Skewness and Kurtosis**

## 📌 What is Skewness?

**Skewness** is a measure of the **asymmetry** of a data distribution.📈

Skewness refers to the degree of asymmetry in a distribution. A positive skew indicates that the distribution has a longer tail on the right side, while a negative skew indicates a longer tail on the left side.




Snip20250406_117.png

### 🔄 Types of Skewness:

- **Symmetrical Distribution**:  
  The left and right sides of the histogram are mirror images.  
  🔸 Mean ≈ Median ≈ Mode  
  🔸 Skewness ≈ 0

- **Positive Skew (Right-Skewed)**:  
  The right tail (larger values) is longer.  
  🔸 Mean > Median > Mode  
  🔸 Few high outliers pull the mean to the right  
  🔸 Example: Income distribution

- **Negative Skew (Left-Skewed)**:  
  The left tail (smaller values) is longer.  
  🔸 Mean < Median < Mode  
  🔸 Few low outliers pull the mean to the left  
  🔸 Example: Exam scores where most students score high

image.png

💡 **What Does It Mean?**
1. Positive Skewness > 0 → Tail is longer on the right (right-skewed)

2. Negative Skewness < 0 → Tail is longer on the left (left-skewed)

3. Skewness = 0 → Symmetrical (normal) distribution



####**📌 Why is Skewness Important?**
**Shows the shape of data**

➤ Tells if data is symmetric or has a longer tail on one side.

**Affects mean and median**

➤ In right-skewed data: Mean > Median

➤ In left-skewed data: Mean < Median

**Important for correct analysis**

➤ Many statistical tests assume no skewness.

➤ Skewed data may need special tests or adjustments.

**Helps detect outliers**

➤ Skewness can hint at extreme values that affect results.

**Guides decisions in real life**

➤ In finance, psychology, and science — knowing skewness helps make better choices.

**Used for data transformation**

➤ If data is skewed, transformations like log or square root can make it more normal for modeling.

### 📊 Skewness Value Ranges

| **Skewness Value Range** | **Type of Distribution**      | **Description**                                  |
|--------------------------|-------------------------------|--------------------------------------------------|
| Skewness = 0             | Symmetrical                   | Perfectly balanced on both sides                |
| 0 < Skewness < 1         | Moderately Right-Skewed       | Slightly longer tail on the right               |
| Skewness > 1             | Highly Right-Skewed           | Strong right tail, large outliers on right      |
| −1 < Skewness < 0        | Moderately Left-Skewed        | Slightly longer tail on the left                |
| Skewness < −1            | Highly Left-Skewed            | Strong left tail, large outliers on left        |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Set seed for reproducibility
np.random.seed(42)

# Symmetrical data: normal distribution
symmetrical_data = np.random.normal(loc=50, scale=10, size=1000)

# Right-skewed data: income-like distribution
low_income = np.random.normal(30000, 5000, 900)
high_income = np.random.normal(100000, 20000, 100)
right_skewed_data = np.concatenate((low_income, high_income))
right_skewed_data = right_skewed_data[right_skewed_data > 0]

# Left-skewed data: high exam scores + few low ones
high_scores = np.random.normal(85, 5, 900)
low_scores = np.random.normal(40, 5, 100)
left_skewed_data = np.concatenate((high_scores, low_scores))
left_skewed_data = left_skewed_data[left_skewed_data <= 100]

# Combine all for plotting
datasets = {
    'Symmetrical': symmetrical_data,
    'Left-Skewed': left_skewed_data,
    'Right-Skewed': right_skewed_data
}

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, (label, data) in zip(axes, datasets.items()):
    skew_val = skew(data)
    sns.histplot(data, kde=True, ax=ax, bins=30)
    ax.set_title(f'{label} Distribution\nSkewness = {skew_val:.2f}')
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()

# Output skew values
for name, data in datasets.items():
    print(f"{name} Distribution Skewness: {skew(data):.2f}")
```

# 📌 What is Kurtosis?

**Kurtosis** is a statistical measure that tells us how concentrated the data is around the mean and how heavy or light the tails of a distribution are compared to a normal distribution.

It helps answer:
- Is the data tightly grouped or widely spread?
- Are there many or few outliers?


## 📊 Kurtosis Range

- **Kurtosis = 3** → Normal distribution (Mesokurtic)
- **Kurtosis > 3** → More peaked with heavier tails (Leptokurtic)
- **Kurtosis < 3** → Flatter with lighter tails (Platykurtic)

> 🔎 Excess Kurtosis = Kurtosis - 3  
> It centers the comparison around the normal distribution.

| Kurtosis Value | Excess Kurtosis | Interpretation           |
|----------------|------------------|---------------------------|
| 3              | 0                | Normal (Mesokurtic)       |
| > 3            | > 0              | Heavy tails (Leptokurtic) |
| < 3            | < 0              | Light tails (Platykurtic) |


## 🔄 Types of Kurtosis

### ✅ 1. **Mesokurtic**
- **Kurtosis = 3**
- Normal tail weight and peak
- Moderate chance of outliers
- 📌 Example: Height or weight in a population

### ✅ 2. **Leptokurtic**
- **Kurtosis > 3**
- Sharp peak, heavy tails
- High probability of extreme values (outliers)
- 📌 Example: Stock market returns

### ✅ 3. **Platykurtic**
- **Kurtosis < 3**
- Flat peak, light tails
- Fewer extreme values; more consistency
- 📌 Example: Uniform test scores


## 🎯 Why is Kurtosis Important?

- **Identifies Tail Behavior**  
  Helps understand if your data has a lot of outliers or not.

- **Improves Model Selection**  
  Many statistical models assume normal kurtosis; too high or low may need alternative methods.

- **Supports Risk Assessment**  
  Crucial in finance and quality control where rare extreme events matter.

- **Guides Data Transformation**  
  High kurtosis often requires transformation like log or square root to normalize data.

- **Essential for EDA**  
  Understanding the shape of data improves overall analysis and insight quality.


## ✅ Final Summary

| Kurtosis Type | Kurtosis Value | Outlier Risk | Description                  |
|----------------|----------------|---------------|------------------------------|
| Mesokurtic      | 3              | Moderate      | Normal distribution          |
| Leptokurtic     | > 3            | High          | Tall peak, heavy tails       |
| Platykurtic     | < 3            | Low           | Flat peak, light tails       |

**Analyze Kurtosis in a Pandas DataFrame Column**

```python
import pandas as pd
from scipy.stats import kurtosis

# Example DataFrame: Monthly sales or returns
df = pd.DataFrame({
    'Sales': [220, 250, 270, 400, 600, 620, 1000, 230, 280, 260]
})

kurt = kurtosis(df['Sales'], fisher=False)
if kurt > 3:
    print(f"Kurtosis = {kurt:.2f} → Leptokurtic (heavy tails)")
elif kurt < 3:
    print(f"Kurtosis = {kurt:.2f} → Platykurtic (light tails)")
else:
    print(f"Kurtosis = {kurt:.2f} → Mesokurtic (normal)")
```

# 📌 Z-Score Method for Outlier Detection

The **Z-score method** is a statistical technique used to detect outliers by measuring how far a data point deviates from the mean in terms of standard deviations.


Snip20250406_118.png

## 🔍 How It Works

- If **Z ≈ 0** → Value is close to the mean
- If **Z > 3** or **Z < -3** → Value is far from the mean → **Potential Outlier**

> **Common threshold**:  
> A data point is considered an outlier if its **Z-score > +3** or **Z-score < -3**


## 📊 Example

Given values:  
`[10, 12, 12, 13, 14, 13, 15, 16, 110, 115]`

- Most values lie between 10 and 16  
- **110 and 115** are far from the rest  
- Their Z-scores will be significantly **greater than 3**
- ✅ Marked as **outliers**


## 🧠 When to Use Z-Score

- Best for **normally distributed** (symmetric) data
- Works when your dataset doesn’t already have extreme values skewing the mean


## ⚠️ Limitations

- Not ideal for **skewed or non-normal data**
- Can be **influenced by existing outliers**, which distort the mean and standard deviation


## ✅ Z-Score Interpretation

| Z-Score Range     | Meaning                  |
|-------------------|---------------------------|
| -1 to +1          | Typical / Common value    |
| -2 to -1 / 1 to 2 | Moderate deviation        |
| -3 to -2 / 2 to 3 | Unusual but possible      |
| < -3 or > +3      | **Likely Outlier**        |


## 💡 Tip

Always visualize with **box plots** or **histograms** along with Z-scores to better understand your outliers!


### Example
**Q: In a retail sales dataset, how can we detect unusually high or low daily sales figures using the Z-score method in Python?**

Imagine you're analyzing daily sales for a store, and you notice that some days have extremely high sales — possibly due to holiday promotions, data entry errors, or special events.
To ensure your analysis isn't skewed by these outliers, you want to identify and examine them separately.



```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Step 1: Simulated daily sales data (with high and low outliers)
sales_data = [ 150, 160, 165, 170, 155, 160, 158, 162, 5000, 5200, 60000]

# Step 2: Create DataFrame
df = pd.DataFrame({'Daily_Sales': sales_data})

# Step 3: Calculate Z-scores for each sales value
df['Z_score'] = zscore(df['Daily_Sales'])

# Step 4: Define thresholds for both tails
upper_threshold = 3
lower_threshold = -3

# Step 5: Flag outliers separately
df['Outlier_Type'] = df['Z_score'].apply(
    lambda x: 'High' if x > upper_threshold else ('Low' if x < lower_threshold else 'Normal')
)

# Step 6: Filter out only the outliers
outliers_df = df[df['Outlier_Type'] != 'Normal']

# Step 7: Print the result
print("\nDetected Outliers:\n", outliers_df)
```

**✅ Output Explanation:**
1. Z_score > 3 → High outlier (extremely large value)

2. Z_score < -3 → Low outlier (extremely small or negative value)

3. Between -3 and 3 → Normal

# Correlation and Covariance

## 🔵 Covariance: The Measure of Joint Variation

**Definition:**  
Covariance measures how two random variables change together.

- **Positive Covariance** ➔ As one variable increases, the other tends to increase.
- **Negative Covariance** ➔ As one variable increases, the other tends to decrease.

image.png

**Key Insight:**
- Covariance gives the **direction** of the relationship, but **not the strength**.
- It is sensitive to the **units of measurement** (e.g., dollars, kilograms).

**🔹 Example:**  
If we track study time (hours) and exam scores (marks), and they both tend to rise together, the covariance will be positive!


## 🟠 Correlation: The Scale-Free Relationship

**Definition:**  
Correlation measures both the **strength and direction** of a linear relationship between two variables.

image.png

**Key Insight:**
- Correlation is **unitless** and always between **-1 and 1**.
  - **+1** ➔ Perfect positive linear relationship
  - **-1** ➔ Perfect negative linear relationship
  - **0** ➔ No linear relationship

**🔸 Example:**  
In an e-commerce dataset, the number of items in a cart and the total purchase amount are likely highly positively correlated!

#### **Using the Iris dataset, calculate the covariance and correlation between petal length (cm) and petal width (cm). Provide the code and derive insights from the results.**

```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Display the first few rows
print(iris_df.head())

# Calculate Covariance
covariance = iris_df[['petal length (cm)', 'petal width (cm)']].cov()
print("\nCovariance Matrix:\n", covariance)

# Calculate Correlation
correlation = iris_df[['petal length (cm)', 'petal width (cm)']].corr()
print("\nCorrelation Matrix:\n", correlation)

# Visualize the relationship with a scatter plot
plt.scatter(iris_df['petal length (cm)'], iris_df['petal width (cm)'])
plt.title('Scatter plot of Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()
```

# Insights on Covariance and Correlation

**Covariance Value ≈ 1.296**

- **Positive covariance** indicates that as **petal length** increases, **petal width** also tends to increase.
- The **magnitude (1.296)** is **context-dependent**, influenced by the **units of measurement** in the data.

**Correlation Coefficient ≈ 0.963**

- Indicates a **very strong positive linear relationship** between **petal length** and **petal width**.
- Being **close to +1** suggests that longer petals tend to be wider **almost linearly**.

**Scatter Plot Insights**

- The scatter plot will show points forming an **upward trend**, visually confirming the **strong positive correlation**.
- There is **very little spread** around the trend line, reinforcing the **strength** and **consistency** of the relationship.
