# Loops
A loop in programming is a way to repeat a set of instructions or actions multiple times without writing the same code over and over again. It's like telling the computer:

"Hey, keep doing this task until you're told to stop!"

````python
# Print hello 5 times
print("hello")
print("hello")
print("hello")
print("hello")
print("hello")
````

# For loop
A for loop is a type of loop in programming that repeats a block of code for each item in a collection (like a list,tuple,set,dict, range of numbers, or string). It’s great for when you know how many times you want to repeat an action.

**How Does a For Loop Work?**

The for loop goes through each item in a sequence (e.g., a list or a range) one at a time and runs the code inside the loop for each item.



```
for item  in collection:
  print(item)
```
* collection: list,tuple,set,dict,range
* item : variable which takes each element from collection



**Breaking it Down:**
1. for: The keyword to start the loop.
2. item: A variable name that represents the current element in the collection.
3. in collection: Refers to the list, range, or sequence the loop will iterate over.
4. Indented block: The actions that will be repeated.

# Break
Break in for loop used to stop the loop when certain condition is met.

````python
A = ["red","green","pink","blue","orange","black"]
# stop the loop when it takes blue
for x in A:
  print(x)
  if x=='blue':
    break
print("The loop has ended")
````

````python
B = [100,200,410,200,600]
# Stop the loop when value is greater than 400
for x in B:
  print(x)
  if x>400:
    break
  print(x)
print("loop has ended")
````

# continue
Continue in for loop skips the iteration which satisfies the condition and continue with rest of the iteration.

````python
A = [10 ,5, -2 , 3, -5 ,-6 ,7 ,-8]
# print only positive numbers
for x in A:
  if x<0:
    continue
  print(x)
````

````python
B = [2,3,5,6,7,8,9,10,12]
#B = list((2,3,5,6,7,8,9,10,12))
# print only even numbers
for x in B:
  print("hii")
  if x%2==1:
    continue
  print(x)
  print("abhi")
````

````python
nested_list = [[10, 15, 20], [5, 8, 12], [11, 14, 9]]
# count of odd and even
even_count=0
odd_count=0
for inner_list in nested_list:
  #print(inner_list)
  for x in inner_list:
    #print(x)
    if x%2==0:
      even_count+=1
    else:
      odd_count+=1
print(f"total even count is {even_count} and total odd count is {odd_count}")
````
