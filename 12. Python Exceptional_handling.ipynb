```python
# raise error with negative numbers
age = int(input())
if 0<age<18:
  print("Not eligible to vote")
elif age >=18:
  print("Eligible to vote")
else:
  raise ValueError("Number cannot be less than zero")
```

```python
# raising error with negative numbers and handling them also
age = int(input("Enter your age: "))
try:
  if 0<age<18:
    print("Not eligible to vote")
  elif 18<=age<=100:
    print("Eligible to vote")
  else:
    raise ValueError("Your age should not be negative and greater than 100")
except Exception as e:
  print(type(e).__name__)
  print(e)
```

```python
# raising error with negative numbers and handling them also
age = int(input("Enter your age: "))
try:
  if 0<age<18:
    print("Not eligible to vote")
  elif age >=18:
    print("Eligible to vote")
  else:
    raise ValueError("Number cannot be less than zero Abhishek")
except Exception as e:
  print(type(e).__name__)
  print(e)
```

```python
# raise errors with negative numbers and handling them also with try block
try:
  age = int(input())
  if 0<age<18:
    print("Not eligible to vote")
  elif age >=18:
    print("Eligible to vote")
  else:
    raise ValueError("Number cannot be less than zero")
except Exception as e:
  print(e)
```

```python
# raise errors
age = int(input())
try:
  if 1<age<18:
    print("Not eligible to vote")
  elif age >=18:
    print("Eligible to vote")
  else:
    raise ValueError("Number cannot be less than or equal zero")
except Exception as e:
  print(e)
```

1. **Basic `try` and `except`:** Handle specific exceptions like `ZeroDivisionError` to prevent crashes.
2. **Add `finally`:** Ensure code always runs, even if an error occurs (e.g., for cleanup).
3. **Handle Specific Exceptions:** Use multiple `except` blocks to handle different types of errors like `ValueError` and `ZeroDivisionError`.
4. **Custom Exceptions:** Define and raise your own errors for specific conditions using the `raise` keyword.
