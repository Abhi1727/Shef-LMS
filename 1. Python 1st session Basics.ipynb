# HOW PYTHON WORKS?

**Understanding a Python Compiler Through the Human Body**

Imagine that the process of compiling code in Python is somewhat similar to how our body receives and processes sensory information. Let's use the analogy of our senses (like our eyes and ears) and our brain to understand this.

### **Python Compiler as the Human Body**

Think of your Python program as a human body going through a daily process:

---

### **1. Sensory Input (Eyes and Ears = Your Python Code):**

- Just like your body gets information through your **eyes** (visual input) and **ears** (sound), your Python program starts with **input**—the instructions you write in your Python file.
- Your Python code is like raw data coming in.

---

### **2. The Brain (Processing and Compiling):**

- Your **brain** takes the sensory input (what you see and hear) and **makes sense of it**. For example, when you see a ball flying towards you, your brain quickly decides: “I need to catch this ball.”
- Similarly, Python **compiles** your raw code (written in human-readable text) into **bytecode**—a language that the computer’s “brain” (the processor) understands.

---

### **3. Meaningful Actions (Organized Instructions):**

- Once your brain processes the input, it forms a plan. For example, it might tell your muscles: “Move your hand quickly to the right to catch the ball.”
- Python does the same: It converts your compiled bytecode into **instructions** that the computer can execute step by step.

---

### **4. Execution (Taking Action):**

- Finally, your body acts on the brain’s instructions, like moving your hand to catch the ball.
- Similarly, Python executes the bytecode instructions, performing tasks like showing results on the screen or saving data to a file.

---

### **Key Takeaway:**

- Your **eyes and ears** are like the **Python code input**.
- Your **brain** is the **compiler**, turning raw input into meaningful actions.
- Your **body’s actions** are like the **execution phase**, where tasks get done.

The Python compiler is your program's "brain," translating your thoughts (code) into actions the computer can carry out! 😊

**Wondering what happened in the backend of this code?**

A **compiler** in Python, or any programming language, is like a translator for computers. It takes the human-readable code that a programmer writes and turns it into something the computer can understand and execute.

Here's how it works in simple terms:

1. **Writing Code**: First, a programmer writes code in a high-level language like Python. This code is easy for humans to read and write because it uses words and structures that make sense to us.
2. **Compilation**: When the programmer is ready to run their program, the Python compiler comes into play. The compiler is a special program that reads the Python code and translates it into a lower-level language called "machine code" or "bytecode." This lower-level language is like the computer's native language, and it can be executed very quickly.
3. **Errors and Optimization**: The compiler checks the code for any mistakes or errors. If it finds any, it will report them to the programmer, so they can fix the issues. The compiler can also optimize the code to make it run faster, just like an editor might revise and improve a written text.
4. **Execution**: Once the code is error-free and optimized, it's ready to be executed by the computer. The computer can now understand the translated code and perform the tasks described in the original Python code.

So, in a nutshell, a Python compiler is like a bridge between human-readable code and the computer's language. It takes our instructions, translates them into a format the computer can work with, and ensures everything is correct and efficient before the program runs. It's a crucial part of the software development process, allowing us to turn our ideas into functional software.

# VARIABLES

In programming, a variable is a named storage location that holds a value. Think of it like a container or a box where you can store data. The value stored in a variable can change during the execution of a program.

Variables are essential because they allow us to:

*   **Store data:** Keep track of information that the program needs to use.
*   **Manipulate data:** Perform operations on the stored values.
*   **Make code readable:** Give meaningful names to data, making the code easier to understand.

In Python, you don't need to declare the type of a variable explicitly. Python is dynamically typed, meaning the type is determined at runtime based on the value assigned to the variable.

**Creating Variables:**

You create a variable by assigning a value to a name using the assignment operator (`=`).

```python
name = "Abhishek"
age = 16
salary = 100000000000 # The value of age is updated

print(name)
print(age)
print(salary)
```
